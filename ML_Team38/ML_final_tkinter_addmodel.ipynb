```python
fp = functools.partial
class VerticalScrolledFrame(ttk.Frame):
    """
    A pure Tkinter scrollable frame that actually works! 
    * Use the 'interior' attribute to place widgets inside the scrollable frame
    * Construct and pack/place/grid normally
    * This frame only allows vertical scrolling
    * -- NOTE: You will need to comment / uncomment code the differently for windows or linux 
    * -- or write your own 'os' type check.
    * This comes from a different naming of the the scrollwheel 'button', on different systems.
    """
    def __init__(self, parent, *args, **kw):

        # track changes to the canvas and frame width and sync them,
        # also updating the scrollbar
        def _configure_interior(event):
            # update the scrollbars to match the size of the inner frame
            size = (interior.winfo_reqwidth(), interior.winfo_reqheight())
            canvas.config(scrollregion="0 0 %s %s" % size)
            if interior.winfo_reqwidth() != canvas.winfo_width():
                # update the canvas's width to fit the inner frame
                canvas.config(width=interior.winfo_reqwidth())

        def _configure_canvas(event):
            if interior.winfo_reqwidth() != canvas.winfo_width():
                # update the inner frame's width to fill the canvas
                canvas.itemconfigure(interior_id, width=canvas.winfo_width())

        """
        This is linux code for scrolling the window, 
        It has different buttons for scrolling the windows
        """
        def _on_mousewheel(event, scroll):
            canvas.yview_scroll(int(scroll), "units")

        def _bind_to_mousewheel(event):
            canvas.bind_all("<Button-4>", fp(_on_mousewheel, scroll=-1))
            canvas.bind_all("<Button-5>", fp(_on_mousewheel, scroll=1))

        def _unbind_from_mousewheel(event):
            canvas.unbind_all("<Button-4>")
            canvas.unbind_all("<Button-5>")


        """
        This is windows code for scrolling the Frame
        def _on_mousewheel(event):
            canvas.yview_scroll(int(-1*(event.delta/120)), "units")
        def _bind_to_mousewheel(event):
            canvas.bind_all("<MouseWheel>", _on_mousewheel)
        def _unbind_from_mousewheel(event):
            canvas.unbind_all("<MouseWheel>")
        """

        ttk.Frame.__init__(self, parent, *args, **kw)

        # create a canvas object and a vertical scrollbar for scrolling it
        vscrollbar = ttk.Scrollbar(self, orient=tk.VERTICAL)
        vscrollbar.pack(fill=tk.Y, side=tk.RIGHT, expand=tk.FALSE)
        canvas = tk.Canvas(self, bd=0, highlightthickness=0,
                           yscrollcommand=vscrollbar.set)
        canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=tk.TRUE)
        vscrollbar.config(command=canvas.yview)

        # reset the view
        canvas.xview_moveto(0)
        canvas.yview_moveto(0)

        # create a frame inside the canvas which will be scrolled with it
        self.interior = interior = ttk.Frame(canvas)
        interior_id = canvas.create_window(0, 0, window=interior,
                                           anchor=tk.NW)

        interior.bind('<Configure>', _configure_interior)
        canvas.bind('<Configure>', _configure_canvas)
        canvas.bind('<Enter>', _bind_to_mousewheel)
        canvas.bind('<Leave>', _unbind_from_mousewheel)
#---------------------------------------------------------------
root = tk.Tk()   # 建立 tkinter 視窗物件
root.title('Hogwarts_Sorting_Hat')        # 設定標題
# root.iconbitmap('favicon.ico')  # 設定 icon ( 格式限定 .ico )
window_width = root.winfo_screenwidth()    # 取得螢幕寬度
window_height = root.winfo_screenheight()  # 取得螢幕高度
width = 800
height = 600
left = int((window_width - width)/2)       # 計算左上 x 座標
top = int((window_height - height)/2)      # 計算左上 y 座標
root.geometry(f'{width}x{height}+{left}+{top}')
# ------------------------------------------------------------------------
canvas = tk.Canvas(root, width=800, height=50)
canvas.create_text(400, 20, text='Welcome to Wizarding World', anchor='center', font=('Consolas', 20))
canvas.pack()
# ------------------------------------------------------------------------
def show(*e):
    global Sorting_house_wish, occupation
    Sorting_house_wish = Sorting_house_wish_value.get()
    Sorting_house_wish_a.set('Sorting_house_wish: ' + Sorting_house_wish_value.get())     # Label 變數改變成選單內容，使用 get() 取值
    age = age_scale_h.get()
    age_a.set(f'age: {age_scale_h.get()}')
    occupation = occupation_value.get()
    occupation_a.set('occupation: ' + occupation_value.get())     # Label 變數改變成選單內容，使用 get() 取值
    gender = gender_value.get()
    gender_a.set('gender: ' + gender_value.get())     # Label 變數改變成選單內容，使用 get() 取值
# ------------------------------------------------------------------------
#add new frame 
tframe = tk.Frame(root, width=800, height=100)
tframe.pack(fill='both',expand=tk.TRUE)
total_frame = VerticalScrolledFrame(tframe)
total_frame.pack(fill = 'both',expand=tk.TRUE)
#------------------------------------------------------------------------
Sorting_house_frame = tk.Frame(total_frame.interior, pady=5, padx=5)                  # 加入 Frame 框架
Sorting_house_frame.pack(fill = 'x')
Sorting_house_wish = 'Gryffindor'
Sorting_house_wish_a = tk.StringVar()  # 設定 a 為文字變數
Sorting_house_wish_a.set('Sorting_house_wish: ' + Sorting_house_wish)    # 設定 a 的內容

Sorting_house_wish_label = tk.Label(Sorting_house_frame, textvariable=Sorting_house_wish_a, font=('Consolas',12))  # 放入標籤
Sorting_house_wish_label.pack(side='left')

Sorting_house_wish_list = ['Gryffindor','Ravenclaw','Hufflepuff','Slytherin']   # 選項
Sorting_house_wish_value = tk.StringVar()                                        # 取值
Sorting_house_wish_value.set('Gryffindor')

Sorting_house_wish_menu = tk.OptionMenu(Sorting_house_frame, Sorting_house_wish_value, *Sorting_house_wish_list)     # 第二個參數是取值，第三個開始是選項，使用星號展開
Sorting_house_wish_menu.config(width=10, fg='#000')                # 設定樣式
Sorting_house_wish_menu.pack(padx=5, side='left')
Sorting_house_wish_value.trace('w', show)                          # 變數 trace 是否改變，若有改變執行 show
# ------------------------------------------------------------------------
age_frame = tk.Frame(total_frame.interior, pady=5, padx=5)                  # 加入 Frame 框架
age_frame.pack(fill = 'x')
age = '10'
age_a = tk.StringVar()  # 設定 a 為文字變數
age_a.set('age: ' + str(age))    # 設定 a 的內容

age_label = tk.Label(age_frame, textvariable=age_a, font=('Consolas',12))  # 放入標籤
age_label.pack(side='left')

age_scale_h = tk.Scale(age_frame, from_=10, to=60, orient='horizontal', command=show)   # 加入水平調整滑桿 ( orient='horizontal' )
age_scale_h.pack(padx=5, side='left')
# ------------------------------------------------------------------------
occupation_frame = tk.Frame(total_frame.interior, pady=5, padx=5)                  # 加入 Frame 框架
occupation_frame.pack(fill = 'x')
occupation = 'Student'
occupation_a = tk.StringVar()  # 設定 a 為文字變數
occupation_a.set('occupation: ' + occupation)    # 設定 a 的內容

occupation_label = tk.Label(occupation_frame, textvariable=occupation_a, font=('Consolas',12))  # 放入標籤
occupation_label.pack(side='left')

occupation_list = ['Student','Employed','Unemployed','Others']   # 選項
occupation_value = tk.StringVar()                                        # 取值
occupation_value.set('Student')

occupation_menu = tk.OptionMenu(occupation_frame, occupation_value, *occupation_list)     # 第二個參數是取值，第三個開始是選項，使用星號展開
occupation_menu.config(width=10, fg='#000')                # 設定樣式
occupation_menu.pack(padx=5, side='left')
occupation_value.trace('w', show)                          # 變數 trace 是否改變，若有改變執行 show
# ------------------------------------------------------------------------
gender_frame = tk.Frame(total_frame.interior, pady=5, padx=5)                  # 加入 Frame 框架
gender_frame.pack(fill = 'x')
gender = 'Female'
gender_a = tk.StringVar()  # 設定 a 為文字變數
gender_a.set('gender: ' + gender)    # 設定 a 的內容

gender_label = tk.Label(gender_frame, textvariable=gender_a, font=('Consolas',12))  # 放入標籤
gender_label.pack(side='left')

gender_list = ['Female','Male']   # 選項
gender_value = tk.StringVar()                                        # 取值
gender_value.set('Female')

gender_menu = tk.OptionMenu(gender_frame, gender_value, *gender_list)     # 第二個參數是取值，第三個開始是選項，使用星號展開
gender_menu.config(width=10, fg='#000')                # 設定樣式
gender_menu.pack(padx=5, side='left')
gender_value.trace('w', show)                          # 變數 trace 是否改變，若有改變執行 show
# ------------------------------------------------------------------------
SD3_M_frame = tk.Frame(total_frame.interior, pady=5, padx=5)                  # 加入 Frame 框架
SD3_M_frame.pack(fill = 'both')
SD3_M_canvas = tk.Canvas(SD3_M_frame, width=800, height=50)
SD3_M_canvas.create_text(400, 20, text='Short Dark Triad (SD3) Machiavellianism', anchor='center', font=('Consolas', 14))
SD3_M_canvas.create_text(400, 40, text='Strongly disagree = 1, Disagree = 2, Neither agree nor disagree = 3, Agree = 4, Strongly agree = 5', anchor='center', font=('Consolas', 10))
SD3_M_canvas.pack(pady=5) 
# ------------------------------------------------------------------------
SD3_M1_frame = tk.Frame(SD3_M_frame, pady=5)                  # 加入 Frame 框架
SD3_M1_frame.pack(fill = 'x')
SD3_M1_val = tk.StringVar()
SD3_M1_val.set('3')
SD3_M1_a = tk.StringVar()  # 設定 a 為文字變數
SD3_M1_a.set("1. It's not wise to tell your secret.")    # 設定 a 的內容
SD3_M1_label = tk.Label(SD3_M1_frame, textvariable=SD3_M1_a, font=('Consolas',10))  # 放入標籤
SD3_M1_label.pack(side='left')
SD3_M1_label2 = tk.Label(SD3_M1_frame, textvariable=SD3_M1_val, font=('Consolas',10))  # 放入標籤
SD3_M1_label2.pack(side='left')
SD3_M1_radio_btn1 = tk.Radiobutton(SD3_M1_frame, variable=SD3_M1_val, text='1', value='1')    # 放入第一個單選按鈕
SD3_M1_radio_btn1.pack(padx=1, side='left')
SD3_M1_radio_btn2 = tk.Radiobutton(SD3_M1_frame, variable=SD3_M1_val, text='2', value='2')    # 放入第一個單選按鈕
SD3_M1_radio_btn2.pack(padx=1, side='left')
SD3_M1_radio_btn3 = tk.Radiobutton(SD3_M1_frame, variable=SD3_M1_val, text='3', value='3')    # 放入第一個單選按鈕
SD3_M1_radio_btn3.pack(padx=1, side='left')
SD3_M1_radio_btn4 = tk.Radiobutton(SD3_M1_frame, variable=SD3_M1_val, text='4', value='4')    # 放入第一個單選按鈕
SD3_M1_radio_btn4.pack(padx=1, side='left')
SD3_M1_radio_btn5 = tk.Radiobutton(SD3_M1_frame, variable=SD3_M1_val, text='5', value='5')    # 放入第一個單選按鈕
SD3_M1_radio_btn5.pack(padx=1, side='left')
#---
SD3_M2_frame = tk.Frame(SD3_M_frame, pady=5)                  # 加入 Frame 框架
SD3_M2_frame.pack(fill = 'x')
SD3_M2_val = tk.StringVar()
SD3_M2_val.set('3')
SD3_M2_a = tk.StringVar()  # 設定 a 為文字變數
SD3_M2_a.set("2. I like to use clever manipulation to get my way.")    # 設定 a 的內容
SD3_M2_label = tk.Label(SD3_M2_frame, textvariable=SD3_M2_a, font=('Consolas',10))  # 放入標籤
SD3_M2_label.pack(side='left')
SD3_M2_label2 = tk.Label(SD3_M2_frame, textvariable=SD3_M2_val, font=('Consolas',10))  # 放入標籤
SD3_M2_label2.pack(side='left')
SD3_M2_radio_btn1 = tk.Radiobutton(SD3_M2_frame, text='1', variable=SD3_M2_val, value='1')    # 放入第一個單選按鈕
SD3_M2_radio_btn1.pack(padx=1, side='left')
SD3_M2_radio_btn2 = tk.Radiobutton(SD3_M2_frame, text='2', variable=SD3_M2_val, value='2')    # 放入第一個單選按鈕
SD3_M2_radio_btn2.pack(padx=1, side='left')
SD3_M2_radio_btn3 = tk.Radiobutton(SD3_M2_frame, text='3', variable=SD3_M2_val, value='3')    # 放入第一個單選按鈕
SD3_M2_radio_btn3.pack(padx=1, side='left')
SD3_M2_radio_btn4 = tk.Radiobutton(SD3_M2_frame, text='4', variable=SD3_M2_val, value='4')    # 放入第一個單選按鈕
SD3_M2_radio_btn4.pack(padx=1, side='left')
SD3_M2_radio_btn5 = tk.Radiobutton(SD3_M2_frame, text='5', variable=SD3_M2_val, value='5')    # 放入第一個單選按鈕
SD3_M2_radio_btn5.pack(padx=1, side='left')
#---
SD3_M3_frame = tk.Frame(SD3_M_frame, pady=5)                  # 加入 Frame 框架
SD3_M3_frame.pack(fill = 'x')
SD3_M3_val = tk.StringVar()
SD3_M3_val.set('3')
SD3_M3_a = tk.StringVar()  # 設定 a 為文字變數
SD3_M3_a.set("3. Whatever it takes, you must get the important people on your side.")    # 設定 a 的內容
SD3_M3_label = tk.Label(SD3_M3_frame, textvariable=SD3_M3_a, font=('Consolas',10))  # 放入標籤
SD3_M3_label.pack(side='left')
SD3_M3_label2 = tk.Label(SD3_M3_frame, textvariable=SD3_M3_val, font=('Consolas',10))  # 放入標籤
SD3_M3_label2.pack(side='left')
SD3_M3_radio_btn1 = tk.Radiobutton(SD3_M3_frame, text='1', variable=SD3_M3_val, value='1')    # 放入第一個單選按鈕
SD3_M3_radio_btn1.pack(padx=1, side='left')
SD3_M3_radio_btn2 = tk.Radiobutton(SD3_M3_frame, text='2', variable=SD3_M3_val, value='2')    # 放入第一個單選按鈕
SD3_M3_radio_btn2.pack(padx=1, side='left')
SD3_M3_radio_btn3 = tk.Radiobutton(SD3_M3_frame, text='3', variable=SD3_M3_val, value='3')    # 放入第一個單選按鈕
SD3_M3_radio_btn3.pack(padx=1, side='left')
SD3_M3_radio_btn4 = tk.Radiobutton(SD3_M3_frame, text='4', variable=SD3_M3_val, value='4')    # 放入第一個單選按鈕
SD3_M3_radio_btn4.pack(padx=1, side='left')
SD3_M3_radio_btn5 = tk.Radiobutton(SD3_M3_frame, text='5', variable=SD3_M3_val, value='5')    # 放入第一個單選按鈕
SD3_M3_radio_btn5.pack(padx=1, side='left')
#---
SD3_M4_frame = tk.Frame(SD3_M_frame, pady=5)                  # 加入 Frame 框架
SD3_M4_frame.pack(fill = 'x')
SD3_M4_val = tk.StringVar()
SD3_M4_val.set('3')
SD3_M4_a = tk.StringVar()  # 設定 a 為文字變數
SD3_M4_a.set("4. Avoid direct conflict with others because they may be useful in the future.")    # 設定 a 的內容
SD3_M4_label = tk.Label(SD3_M4_frame, textvariable=SD3_M4_a, font=('Consolas',10))  # 放入標籤
SD3_M4_label.pack(side='left')
SD3_M4_label2 = tk.Label(SD3_M4_frame, textvariable=SD3_M4_val, font=('Consolas',10))  # 放入標籤
SD3_M4_label2.pack(side='left')
SD3_M4_radio_btn1 = tk.Radiobutton(SD3_M4_frame, text='1', variable=SD3_M4_val, value='1')    # 放入第一個單選按鈕
SD3_M4_radio_btn1.pack(padx=1, side='left')
SD3_M4_radio_btn2 = tk.Radiobutton(SD3_M4_frame, text='2', variable=SD3_M4_val, value='2')    # 放入第一個單選按鈕
SD3_M4_radio_btn2.pack(padx=1, side='left')
SD3_M4_radio_btn3 = tk.Radiobutton(SD3_M4_frame, text='3', variable=SD3_M4_val, value='3')    # 放入第一個單選按鈕
SD3_M4_radio_btn3.pack(padx=1, side='left')
SD3_M4_radio_btn4 = tk.Radiobutton(SD3_M4_frame, text='4', variable=SD3_M4_val, value='4')    # 放入第一個單選按鈕
SD3_M4_radio_btn4.pack(padx=1, side='left')
SD3_M4_radio_btn5 = tk.Radiobutton(SD3_M4_frame, text='5', variable=SD3_M4_val, value='5')    # 放入第一個單選按鈕
SD3_M4_radio_btn5.pack(padx=1, side='left')
#---
SD3_M5_frame = tk.Frame(SD3_M_frame, pady=5)                  # 加入 Frame 框架
SD3_M5_frame.pack(fill = 'x')
SD3_M5_val = tk.StringVar()
SD3_M5_a = tk.StringVar()  # 設定 a 為文字變數
SD3_M5_val.set('3')
SD3_M5_a.set("5. It’s wise to keep track of information that you can use against people later.")    # 設定 a 的內容
SD3_M5_label = tk.Label(SD3_M5_frame, textvariable=SD3_M5_a, font=('Consolas',10))  # 放入標籤
SD3_M5_label.pack(side='left')
SD3_M5_label2 = tk.Label(SD3_M5_frame, textvariable=SD3_M5_val, font=('Consolas',10))  # 放入標籤
SD3_M5_label2.pack(side='left')
SD3_M5_radio_btn1 = tk.Radiobutton(SD3_M5_frame, text='1', variable=SD3_M5_val, value='1')    # 放入第一個單選按鈕
SD3_M5_radio_btn1.pack(padx=1, side='left')
SD3_M5_radio_btn2 = tk.Radiobutton(SD3_M5_frame, text='2', variable=SD3_M5_val, value='2')    # 放入第一個單選按鈕
SD3_M5_radio_btn2.pack(padx=1, side='left')
SD3_M5_radio_btn3 = tk.Radiobutton(SD3_M5_frame, text='3', variable=SD3_M5_val, value='3')    # 放入第一個單選按鈕
SD3_M5_radio_btn3.pack(padx=1, side='left')
SD3_M5_radio_btn4 = tk.Radiobutton(SD3_M5_frame, text='4', variable=SD3_M5_val, value='4')    # 放入第一個單選按鈕
SD3_M5_radio_btn4.pack(padx=1, side='left')
SD3_M5_radio_btn5 = tk.Radiobutton(SD3_M5_frame, text='5', variable=SD3_M5_val, value='5')    # 放入第一個單選按鈕
SD3_M5_radio_btn5.pack(padx=1, side='left')
#---
SD3_M6_frame = tk.Frame(SD3_M_frame, pady=5)                  # 加入 Frame 框架
SD3_M6_frame.pack(fill = 'x')
SD3_M6_val = tk.StringVar()
SD3_M6_val.set('3')
SD3_M6_a = tk.StringVar()  # 設定 a 為文字變數
SD3_M6_a.set("6. You should wait for the right time to get back at people.")    # 設定 a 的內容
SD3_M6_label = tk.Label(SD3_M6_frame, textvariable=SD3_M6_a, font=('Consolas',10))  # 放入標籤
SD3_M6_label.pack(side='left')
SD3_M6_label2 = tk.Label(SD3_M6_frame, textvariable=SD3_M6_val, font=('Consolas',10))  # 放入標籤
SD3_M6_label2.pack(side='left')
SD3_M6_radio_btn1 = tk.Radiobutton(SD3_M6_frame, text='1', variable=SD3_M6_val, value='1')    # 放入第一個單選按鈕
SD3_M6_radio_btn1.pack(padx=1, side='left')
SD3_M6_radio_btn2 = tk.Radiobutton(SD3_M6_frame, text='2', variable=SD3_M6_val, value='2')    # 放入第一個單選按鈕
SD3_M6_radio_btn2.pack(padx=1, side='left')
SD3_M6_radio_btn3 = tk.Radiobutton(SD3_M6_frame, text='3', variable=SD3_M6_val, value='3')    # 放入第一個單選按鈕
SD3_M6_radio_btn3.pack(padx=1, side='left')
SD3_M6_radio_btn4 = tk.Radiobutton(SD3_M6_frame, text='4', variable=SD3_M6_val, value='4')    # 放入第一個單選按鈕
SD3_M6_radio_btn4.pack(padx=1, side='left')
SD3_M6_radio_btn5 = tk.Radiobutton(SD3_M6_frame, text='5', variable=SD3_M6_val, value='5')    # 放入第一個單選按鈕
SD3_M6_radio_btn5.pack(padx=1, side='left')
#---
SD3_M7_frame = tk.Frame(SD3_M_frame, pady=5)                  # 加入 Frame 框架
SD3_M7_frame.pack(fill = 'x')
SD3_M7_val = tk.StringVar()
SD3_M7_val.set('3')
SD3_M7_a = tk.StringVar()  # 設定 a 為文字變數
SD3_M7_a.set("7. There are things you should hide from other people to preserve your reputation.")    # 設定 a 的內容
SD3_M7_label = tk.Label(SD3_M7_frame, textvariable=SD3_M7_a, font=('Consolas',10))  # 放入標籤
SD3_M7_label.pack(side='left')
SD3_M7_label2 = tk.Label(SD3_M7_frame, textvariable=SD3_M7_val, font=('Consolas',10))  # 放入標籤
SD3_M7_label2.pack(side='left')
SD3_M7_radio_btn1 = tk.Radiobutton(SD3_M7_frame, text='1', variable=SD3_M7_val, value='1')    # 放入第一個單選按鈕
SD3_M7_radio_btn1.pack(padx=1, side='left')
SD3_M7_radio_btn2 = tk.Radiobutton(SD3_M7_frame, text='2', variable=SD3_M7_val, value='2')    # 放入第一個單選按鈕
SD3_M7_radio_btn2.pack(padx=1, side='left')
SD3_M7_radio_btn3 = tk.Radiobutton(SD3_M7_frame, text='3', variable=SD3_M7_val, value='3')    # 放入第一個單選按鈕
SD3_M7_radio_btn3.pack(padx=1, side='left')
SD3_M7_radio_btn4 = tk.Radiobutton(SD3_M7_frame, text='4', variable=SD3_M7_val, value='4')    # 放入第一個單選按鈕
SD3_M7_radio_btn4.pack(padx=1, side='left')
SD3_M7_radio_btn5 = tk.Radiobutton(SD3_M7_frame, text='5', variable=SD3_M7_val, value='5')    # 放入第一個單選按鈕
SD3_M7_radio_btn5.pack(padx=1, side='left')
#---
SD3_M8_frame = tk.Frame(SD3_M_frame, pady=5)                  # 加入 Frame 框架
SD3_M8_frame.pack(fill = 'x')
SD3_M8_val = tk.StringVar()
SD3_M8_val.set('3')
SD3_M8_a = tk.StringVar()  # 設定 a 為文字變數
SD3_M8_a.set("8. Make sure your plans benefit yourself, not others.")    # 設定 a 的內容
SD3_M8_label = tk.Label(SD3_M8_frame, textvariable=SD3_M8_a, font=('Consolas',10))  # 放入標籤
SD3_M8_label.pack(side='left')
SD3_M8_label2 = tk.Label(SD3_M8_frame, textvariable=SD3_M8_val, font=('Consolas',10))  # 放入標籤
SD3_M8_label2.pack(side='left')
SD3_M8_radio_btn1 = tk.Radiobutton(SD3_M8_frame, text='1', variable=SD3_M8_val, value='1')    # 放入第一個單選按鈕
SD3_M8_radio_btn1.pack(padx=1, side='left')
SD3_M8_radio_btn2 = tk.Radiobutton(SD3_M8_frame, text='2', variable=SD3_M8_val, value='2')    # 放入第一個單選按鈕
SD3_M8_radio_btn2.pack(padx=1, side='left')
SD3_M8_radio_btn3 = tk.Radiobutton(SD3_M8_frame, text='3', variable=SD3_M8_val, value='3')    # 放入第一個單選按鈕
SD3_M8_radio_btn3.pack(padx=1, side='left')
SD3_M8_radio_btn4 = tk.Radiobutton(SD3_M8_frame, text='4', variable=SD3_M8_val, value='4')    # 放入第一個單選按鈕
SD3_M8_radio_btn4.pack(padx=1, side='left')
SD3_M8_radio_btn5 = tk.Radiobutton(SD3_M8_frame, text='5', variable=SD3_M8_val, value='5')    # 放入第一個單選按鈕
SD3_M8_radio_btn5.pack(padx=1, side='left')
#---
SD3_M9_frame = tk.Frame(SD3_M_frame, pady=5)                  # 加入 Frame 框架
SD3_M9_frame.pack(fill = 'x')
SD3_M9_val = tk.StringVar()
SD3_M9_val.set('3')
SD3_M9_a = tk.StringVar()  # 設定 a 為文字變數
SD3_M9_a.set("9. Most people can be manipulated.")    # 設定 a 的內容
SD3_M9_label = tk.Label(SD3_M9_frame, textvariable=SD3_M9_a, font=('Consolas',10))  # 放入標籤
SD3_M9_label.pack(side='left')
SD3_M9_label2 = tk.Label(SD3_M9_frame, textvariable=SD3_M9_val, font=('Consolas',10))  # 放入標籤
SD3_M9_label2.pack(side='left')
SD3_M9_radio_btn1 = tk.Radiobutton(SD3_M9_frame, text='1', variable=SD3_M9_val, value='1')    # 放入第一個單選按鈕
SD3_M9_radio_btn1.pack(padx=1, side='left')
SD3_M9_radio_btn2 = tk.Radiobutton(SD3_M9_frame, text='2', variable=SD3_M9_val, value='2')    # 放入第一個單選按鈕
SD3_M9_radio_btn2.pack(padx=1, side='left')
SD3_M9_radio_btn3 = tk.Radiobutton(SD3_M9_frame, text='3', variable=SD3_M9_val, value='3')    # 放入第一個單選按鈕
SD3_M9_radio_btn3.pack(padx=1, side='left')
SD3_M9_radio_btn4 = tk.Radiobutton(SD3_M9_frame, text='4', variable=SD3_M9_val, value='4')    # 放入第一個單選按鈕
SD3_M9_radio_btn4.pack(padx=1, side='left')
SD3_M9_radio_btn5 = tk.Radiobutton(SD3_M9_frame, text='5', variable=SD3_M9_val, value='5')    # 放入第一個單選按鈕
SD3_M9_radio_btn5.pack(padx=1, side='left')
#---
# ------------------------------------------------------------------------
SD3_N_frame = tk.Frame(total_frame.interior, pady=5, padx=5)                  # 加入 Frame 框架 (Narcissism)
SD3_N_frame.pack(fill = 'both')
canvas = tk.Canvas(SD3_N_frame, width=800, height=50)
canvas.create_text(400, 20, text='Short Dark Triad (SD3) Narcissism', anchor='center', font=('Consolas', 14))
canvas.create_text(400, 40, text='Strongly disagree = 1, Disagree = 2, Neither agree nor disagree = 3, Agree = 4, Strongly agree = 5', anchor='center', font=('Consolas', 10))
canvas.pack(pady=5)
# sb = tk.Scrollbar(SD3_N_frame)               
# sb.pack(side='right',fill='y') 
# ------------------------------------------------------------------------
SD3_N1_frame = tk.Frame(SD3_N_frame, pady=5)                  # 加入 Frame 框架
SD3_N1_frame.pack(fill = 'x')
SD3_N1_val = tk.StringVar()
SD3_N1_val.set('3')
SD3_N1_a = tk.StringVar()  # 設定 a 為文字變數
SD3_N1_a.set("1. People see me as a natural leader.")    # 設定 a 的內容
SD3_N1_label = tk.Label(SD3_N1_frame, textvariable=SD3_N1_a, font=('Consolas',10))  # 放入標籤
SD3_N1_label.pack(side='left')
SD3_N1_label2 = tk.Label(SD3_N1_frame, textvariable=SD3_N1_val, font=('Consolas',10))  # 放入標籤
SD3_N1_label2.pack(side='left')
SD3_N1_radio_btn1 = tk.Radiobutton(SD3_N1_frame, text='1', variable=SD3_M9_val, value='1')    # 放入第一個單選按鈕
SD3_N1_radio_btn1.pack(padx=1, side='left')
SD3_N1_radio_btn2 = tk.Radiobutton(SD3_N1_frame, text='2', variable=SD3_N1_val, value='2')    # 放入第一個單選按鈕
SD3_N1_radio_btn2.pack(padx=1, side='left')
SD3_N1_radio_btn3 = tk.Radiobutton(SD3_N1_frame, text='3', variable=SD3_N1_val, value='3')    # 放入第一個單選按鈕
SD3_N1_radio_btn3.pack(padx=1, side='left')
SD3_N1_radio_btn4 = tk.Radiobutton(SD3_N1_frame, text='4', variable=SD3_N1_val, value='4')    # 放入第一個單選按鈕
SD3_N1_radio_btn4.pack(padx=1, side='left')
SD3_N1_radio_btn5 = tk.Radiobutton(SD3_N1_frame, text='5', variable=SD3_N1_val, value='5')    # 放入第一個單選按鈕
SD3_N1_radio_btn5.pack(padx=1, side='left')
#---
SD3_N2_frame = tk.Frame(SD3_N_frame, pady=5)                  # 加入 Frame 框架
SD3_N2_frame.pack(fill = 'x')
SD3_N2_val = tk.StringVar()
SD3_N2_val.set('3')
SD3_N2_a = tk.StringVar()  # 設定 a 為文字變數
SD3_N2_a.set("2. I hate being the center of attention. (R)")    # 設定 a 的內容
SD3_N2_label = tk.Label(SD3_N2_frame, textvariable=SD3_N2_a, font=('Consolas',10))  # 放入標籤
SD3_N2_label.pack(side='left')
SD3_N2_label2 = tk.Label(SD3_N2_frame, textvariable=SD3_N2_val, font=('Consolas',10))  # 放入標籤
SD3_N2_label2.pack(side='left')
SD3_N2_radio_btn1 = tk.Radiobutton(SD3_N2_frame, text='1', variable=SD3_N2_val, value='1')    # 放入第一個單選按鈕
SD3_N2_radio_btn1.pack(padx=1, side='left')
SD3_N2_radio_btn2 = tk.Radiobutton(SD3_N2_frame, text='2', variable=SD3_N2_val, value='2')    # 放入第一個單選按鈕
SD3_N2_radio_btn2.pack(padx=1, side='left')
SD3_N2_radio_btn3 = tk.Radiobutton(SD3_N2_frame, text='3', variable=SD3_N2_val, value='3')    # 放入第一個單選按鈕
SD3_N2_radio_btn3.pack(padx=1, side='left')
SD3_N2_radio_btn4 = tk.Radiobutton(SD3_N2_frame, text='4', variable=SD3_N2_val, value='4')    # 放入第一個單選按鈕
SD3_N2_radio_btn4.pack(padx=1, side='left')
SD3_N2_radio_btn5 = tk.Radiobutton(SD3_N2_frame, text='5', variable=SD3_N2_val, value='5')    # 放入第一個單選按鈕
SD3_N2_radio_btn5.pack(padx=1, side='left')
#---
SD3_N3_frame = tk.Frame(SD3_N_frame, pady=5)                  # 加入 Frame 框架
SD3_N3_frame.pack(fill = 'x')
SD3_N3_val = tk.StringVar()
SD3_N3_val.set('3')
SD3_N3_a = tk.StringVar()  # 設定 a 為文字變數
SD3_N3_a.set("3. Many group activities tend to be dull without me.")    # 設定 a 的內容
SD3_N3_label = tk.Label(SD3_N3_frame, textvariable=SD3_N3_a, font=('Consolas',10))  # 放入標籤
SD3_N3_label.pack(side='left')
SD3_N3_label2 = tk.Label(SD3_N3_frame, textvariable=SD3_N3_val, font=('Consolas',10))  # 放入標籤
SD3_N3_label2.pack(side='left')
SD3_N3_radio_btn1 = tk.Radiobutton(SD3_N3_frame, text='1', variable=SD3_N3_val, value='1')    # 放入第一個單選按鈕
SD3_N3_radio_btn1.pack(padx=1, side='left')
SD3_N3_radio_btn2 = tk.Radiobutton(SD3_N3_frame, text='2', variable=SD3_N3_val, value='2')    # 放入第一個單選按鈕
SD3_N3_radio_btn2.pack(padx=1, side='left')
SD3_N3_radio_btn3 = tk.Radiobutton(SD3_N3_frame, text='3', variable=SD3_N3_val, value='3')    # 放入第一個單選按鈕
SD3_N3_radio_btn3.pack(padx=1, side='left')
SD3_N3_radio_btn4 = tk.Radiobutton(SD3_N3_frame, text='4', variable=SD3_N3_val, value='4')    # 放入第一個單選按鈕
SD3_N3_radio_btn4.pack(padx=1, side='left')
SD3_N3_radio_btn5 = tk.Radiobutton(SD3_N3_frame, text='5', variable=SD3_N3_val, value='5')    # 放入第一個單選按鈕
SD3_N3_radio_btn5.pack(padx=1, side='left')
#---
SD3_N4_frame = tk.Frame(SD3_N_frame, pady=5)                  # 加入 Frame 框架
SD3_N4_frame.pack(fill = 'x')
SD3_N4_val = tk.StringVar()
SD3_N4_val.set('3')
SD3_N4_a = tk.StringVar()  # 設定 a 為文字變數
SD3_N4_a.set("4. I know that I am special because everyone keeps telling me so.")    # 設定 a 的內容
SD3_N4_label = tk.Label(SD3_N4_frame, textvariable=SD3_N4_a, font=('Consolas',10))  # 放入標籤
SD3_N4_label.pack(side='left')
SD3_N4_label2 = tk.Label(SD3_N4_frame, textvariable=SD3_N4_val, font=('Consolas',10))  # 放入標籤
SD3_N4_label2.pack(side='left')
SD3_N4_radio_btn1 = tk.Radiobutton(SD3_N4_frame, text='1', variable=SD3_N4_val, value='1')    # 放入第一個單選按鈕
SD3_N4_radio_btn1.pack(padx=1, side='left')
SD3_N4_radio_btn2 = tk.Radiobutton(SD3_N4_frame, text='2', variable=SD3_N4_val, value='2')    # 放入第一個單選按鈕
SD3_N4_radio_btn2.pack(padx=1, side='left')
SD3_N4_radio_btn3 = tk.Radiobutton(SD3_N4_frame, text='3', variable=SD3_N4_val, value='3')    # 放入第一個單選按鈕
SD3_N4_radio_btn3.pack(padx=1, side='left')
SD3_N4_radio_btn4 = tk.Radiobutton(SD3_N4_frame, text='4', variable=SD3_N4_val, value='4')    # 放入第一個單選按鈕
SD3_N4_radio_btn4.pack(padx=1, side='left')
SD3_N4_radio_btn5 = tk.Radiobutton(SD3_N4_frame, text='5', variable=SD3_N4_val, value='5')    # 放入第一個單選按鈕
SD3_N4_radio_btn5.pack(padx=1, side='left')
#---
SD3_N5_frame = tk.Frame(SD3_N_frame, pady=5)                  # 加入 Frame 框架
SD3_N5_frame.pack(fill = 'x')
SD3_N5_val = tk.StringVar()
SD3_N5_val.set('3')
SD3_N5_a = tk.StringVar()  # 設定 a 為文字變數
SD3_N5_a.set("5. I like to get acquainted with important people.")    # 設定 a 的內容
SD3_N5_label = tk.Label(SD3_N5_frame, textvariable=SD3_N5_a, font=('Consolas',10))  # 放入標籤
SD3_N5_label.pack(side='left')
SD3_N5_label2 = tk.Label(SD3_N5_frame, textvariable=SD3_N5_val, font=('Consolas',10))  # 放入標籤
SD3_N5_label2.pack(side='left')
SD3_N5_radio_btn1 = tk.Radiobutton(SD3_N5_frame, text='1', variable=SD3_N5_val, value='1')    # 放入第一個單選按鈕
SD3_N5_radio_btn1.pack(padx=1, side='left')
SD3_N5_radio_btn2 = tk.Radiobutton(SD3_N5_frame, text='2', variable=SD3_N5_val, value='2')    # 放入第一個單選按鈕
SD3_N5_radio_btn2.pack(padx=1, side='left')
SD3_N5_radio_btn3 = tk.Radiobutton(SD3_N5_frame, text='3', variable=SD3_N5_val, value='3')    # 放入第一個單選按鈕
SD3_N5_radio_btn3.pack(padx=1, side='left')
SD3_N5_radio_btn4 = tk.Radiobutton(SD3_N5_frame, text='4', variable=SD3_N5_val, value='4')    # 放入第一個單選按鈕
SD3_N5_radio_btn4.pack(padx=1, side='left')
SD3_N5_radio_btn5 = tk.Radiobutton(SD3_N5_frame, text='5', variable=SD3_N5_val, value='5')    # 放入第一個單選按鈕
SD3_N5_radio_btn5.pack(padx=1, side='left')
#---
SD3_N6_frame = tk.Frame(SD3_N_frame, pady=5)                  # 加入 Frame 框架
SD3_N6_frame.pack(fill = 'x')
SD3_N6_val = tk.StringVar()
SD3_N6_val.set('3')
SD3_N6_a = tk.StringVar()  # 設定 a 為文字變數
SD3_N6_a.set("6. I feel embarrassed if someone compliments me. (R)")    # 設定 a 的內容
SD3_N6_label = tk.Label(SD3_N6_frame, textvariable=SD3_N6_a, font=('Consolas',10))  # 放入標籤
SD3_N6_label.pack(side='left')
SD3_N6_label2 = tk.Label(SD3_N6_frame, textvariable=SD3_N6_val, font=('Consolas',10))  # 放入標籤
SD3_N6_label2.pack(side='left')
SD3_N6_radio_btn1 = tk.Radiobutton(SD3_N6_frame, text='1', variable=SD3_N6_val, value='1')    # 放入第一個單選按鈕
SD3_N6_radio_btn1.pack(padx=1, side='left')
SD3_N6_radio_btn2 = tk.Radiobutton(SD3_N6_frame, text='2', variable=SD3_N6_val, value='2')    # 放入第一個單選按鈕
SD3_N6_radio_btn2.pack(padx=1, side='left')
SD3_N6_radio_btn3 = tk.Radiobutton(SD3_N6_frame, text='3', variable=SD3_N6_val, value='3')    # 放入第一個單選按鈕
SD3_N6_radio_btn3.pack(padx=1, side='left')
SD3_N6_radio_btn4 = tk.Radiobutton(SD3_N6_frame, text='4', variable=SD3_N6_val, value='4')    # 放入第一個單選按鈕
SD3_N6_radio_btn4.pack(padx=1, side='left')
SD3_N6_radio_btn5 = tk.Radiobutton(SD3_N6_frame, text='5', variable=SD3_N6_val, value='5')    # 放入第一個單選按鈕
SD3_N6_radio_btn5.pack(padx=1, side='left')
#---
SD3_N7_frame = tk.Frame(SD3_N_frame, pady=5)                  # 加入 Frame 框架
SD3_N7_frame.pack(fill = 'x')
SD3_N7_val = tk.StringVar()
SD3_N7_val.set('3')
SD3_N7_a = tk.StringVar()  # 設定 a 為文字變數
SD3_N7_a.set("7. I have been compared to famous people.")    # 設定 a 的內容
SD3_N7_label = tk.Label(SD3_N7_frame, textvariable=SD3_N7_a, font=('Consolas',10))  # 放入標籤
SD3_N7_label.pack(side='left')
SD3_N7_radio_btn1 = tk.Radiobutton(SD3_N7_frame, text='1', variable=SD3_N7_val, value='1')    # 放入第一個單選按鈕
SD3_N7_radio_btn1.pack(padx=1, side='left')
SD3_N7_radio_btn2 = tk.Radiobutton(SD3_N7_frame, text='2', variable=SD3_N7_val, value='2')    # 放入第一個單選按鈕
SD3_N7_radio_btn2.pack(padx=1, side='left')
SD3_N7_radio_btn3 = tk.Radiobutton(SD3_N7_frame, text='3', variable=SD3_N7_val, value='3')    # 放入第一個單選按鈕
SD3_N7_radio_btn3.pack(padx=1, side='left')
SD3_N7_radio_btn4 = tk.Radiobutton(SD3_N7_frame, text='4', variable=SD3_N7_val, value='4')    # 放入第一個單選按鈕
SD3_N7_radio_btn4.pack(padx=1, side='left')
SD3_N7_radio_btn5 = tk.Radiobutton(SD3_N7_frame, text='5', variable=SD3_N7_val, value='5')    # 放入第一個單選按鈕
SD3_N7_radio_btn5.pack(padx=1, side='left')
#---
SD3_N8_frame = tk.Frame(SD3_N_frame, pady=5)                  # 加入 Frame 框架
SD3_N8_frame.pack(fill = 'x')
SD3_N8_val = tk.StringVar()
SD3_N8_val.set('3')
SD3_N8_a = tk.StringVar()  # 設定 a 為文字變數
SD3_N8_a.set("8. I am an average person. (R)")    # 設定 a 的內容
SD3_N8_label = tk.Label(SD3_N8_frame, textvariable=SD3_N8_a, font=('Consolas',10))  # 放入標籤
SD3_N8_label.pack(side='left')
SD3_N8_label2 = tk.Label(SD3_N8_frame, textvariable=SD3_N8_val, font=('Consolas',10))  # 放入標籤
SD3_N8_label2.pack(side='left')
SD3_N8_radio_btn1 = tk.Radiobutton(SD3_N8_frame, text='1', variable=SD3_N8_val, value='1')    # 放入第一個單選按鈕
SD3_N8_radio_btn1.pack(padx=1, side='left')
SD3_N8_radio_btn2 = tk.Radiobutton(SD3_N8_frame, text='2', variable=SD3_N8_val, value='2')    # 放入第一個單選按鈕
SD3_N8_radio_btn2.pack(padx=1, side='left')
SD3_N8_radio_btn3 = tk.Radiobutton(SD3_N8_frame, text='3', variable=SD3_N8_val, value='3')    # 放入第一個單選按鈕
SD3_N8_radio_btn3.pack(padx=1, side='left')
SD3_N8_radio_btn4 = tk.Radiobutton(SD3_N8_frame, text='4', variable=SD3_N8_val, value='4')    # 放入第一個單選按鈕
SD3_N8_radio_btn4.pack(padx=1, side='left')
SD3_N8_radio_btn5 = tk.Radiobutton(SD3_N8_frame, text='5', variable=SD3_N8_val, value='5')    # 放入第一個單選按鈕
SD3_N8_radio_btn5.pack(padx=1, side='left')
#---
SD3_N9_frame = tk.Frame(SD3_N_frame, pady=5)                  # 加入 Frame 框架
SD3_N9_frame.pack(fill = 'x')
SD3_N9_val = tk.StringVar()
SD3_N9_val.set('3')
SD3_N9_a = tk.StringVar()  # 設定 a 為文字變數
SD3_N9_a.set("9. I insist on getting the respect I deserve.")    # 設定 a 的內容
SD3_N9_label = tk.Label(SD3_N9_frame, textvariable=SD3_N9_a, font=('Consolas',10))  # 放入標籤
SD3_N9_label.pack(side='left')
SD3_N9_label2 = tk.Label(SD3_N2_frame, textvariable=SD3_N2_val, font=('Consolas',10))  # 放入標籤
SD3_N9_label2.pack(side='left')
SD3_N9_radio_btn1 = tk.Radiobutton(SD3_N9_frame, text='1', variable=SD3_N9_val, value='1')    # 放入第一個單選按鈕
SD3_N9_radio_btn1.pack(padx=1, side='left')
SD3_N9_radio_btn2 = tk.Radiobutton(SD3_N9_frame, text='2', variable=SD3_N9_val, value='2')    # 放入第一個單選按鈕
SD3_N9_radio_btn2.pack(padx=1, side='left')
SD3_N9_radio_btn3 = tk.Radiobutton(SD3_N9_frame, text='3', variable=SD3_N9_val, value='3')    # 放入第一個單選按鈕
SD3_N9_radio_btn3.pack(padx=1, side='left')
SD3_N9_radio_btn4 = tk.Radiobutton(SD3_N9_frame, text='4', variable=SD3_N9_val, value='4')    # 放入第一個單選按鈕
SD3_N9_radio_btn4.pack(padx=1, side='left')
SD3_N9_radio_btn5 = tk.Radiobutton(SD3_N9_frame, text='5', variable=SD3_N9_val, value='5')    # 放入第一個單選按鈕
SD3_N9_radio_btn5.pack(padx=1, side='left')
# ------------------------------------------------------------------------
SD3_P_frame = tk.Frame(total_frame.interior, pady=5, padx=5)                  # 加入 Frame 框架 (Narcissism)
SD3_P_frame.pack(fill = 'both')
canvas = tk.Canvas(SD3_P_frame, width=800, height=50)
canvas.create_text(400, 20, text='Short Dark Triad (SD3) Psychopathy', anchor='center', font=('Consolas', 14))
canvas.create_text(400, 40, text='Strongly disagree = 1, Disagree = 2, Neither agree nor disagree = 3, Agree = 4, Strongly agree = 5', anchor='center', font=('Consolas', 10))
canvas.pack(pady=5)
# ------------------------------------------------------------------------
SD3_P1_frame = tk.Frame(SD3_P_frame, pady=5)                  # 加入 Frame 框架
SD3_P1_frame.pack(fill = 'x')
SD3_P1_val = tk.StringVar()
SD3_P1_val.set('3')
SD3_P1_a = tk.StringVar()  # 設定 a 為文字變數
SD3_P1_a.set("1. I like to get revenge on authorities.")    # 設定 a 的內容
SD3_P1_label = tk.Label(SD3_P1_frame, textvariable=SD3_P1_a, font=('Consolas',10))  # 放入標籤
SD3_P1_label.pack(side='left')
SD3_P1_label2 = tk.Label(SD3_P1_frame, textvariable=SD3_P1_val, font=('Consolas',10))  # 放入標籤
SD3_P1_label2.pack(side='left')
SD3_P1_radio_btn1 = tk.Radiobutton(SD3_P1_frame, text='1', variable=SD3_P1_val, value='1')    # 放入第一個單選按鈕
SD3_P1_radio_btn1.pack(padx=1, side='left')
SD3_P1_radio_btn2 = tk.Radiobutton(SD3_P1_frame, text='2', variable=SD3_P1_val, value='2')    # 放入第一個單選按鈕
SD3_P1_radio_btn2.pack(padx=1, side='left')
SD3_P1_radio_btn3 = tk.Radiobutton(SD3_P1_frame, text='3', variable=SD3_P1_val, value='3')    # 放入第一個單選按鈕
SD3_P1_radio_btn3.pack(padx=1, side='left')
SD3_P1_radio_btn4 = tk.Radiobutton(SD3_P1_frame, text='4', variable=SD3_P1_val, value='4')    # 放入第一個單選按鈕
SD3_P1_radio_btn4.pack(padx=1, side='left')
SD3_P1_radio_btn5 = tk.Radiobutton(SD3_P1_frame, text='5', variable=SD3_P1_val, value='5')    # 放入第一個單選按鈕
SD3_P1_radio_btn5.pack(padx=1, side='left')
#---
SD3_P2_frame = tk.Frame(SD3_P_frame, pady=5)                  # 加入 Frame 框架
SD3_P2_frame.pack(fill = 'x')
SD3_P2_val = tk.StringVar()
SD3_P2_val.set('3')
SD3_P2_a = tk.StringVar()  # 設定 a 為文字變數
SD3_P2_a.set("2. I avoid dangerous situations. (R)")    # 設定 a 的內容
SD3_P2_label = tk.Label(SD3_P2_frame, textvariable=SD3_P2_a, font=('Consolas',10))  # 放入標籤
SD3_P2_label.pack(side='left')
SD3_P2_label2 = tk.Label(SD3_P2_frame, textvariable=SD3_P2_val, font=('Consolas',10))  # 放入標籤
SD3_P2_label2.pack(side='left')
SD3_P2_radio_btn1 = tk.Radiobutton(SD3_P2_frame, text='1', variable=SD3_P2_val, value='1')    # 放入第一個單選按鈕
SD3_P2_radio_btn1.pack(padx=1, side='left')
SD3_P2_radio_btn2 = tk.Radiobutton(SD3_P2_frame, text='2', variable=SD3_P2_val, value='2')    # 放入第一個單選按鈕
SD3_P2_radio_btn2.pack(padx=1, side='left')
SD3_P2_radio_btn3 = tk.Radiobutton(SD3_P2_frame, text='3', variable=SD3_P2_val, value='3')    # 放入第一個單選按鈕
SD3_P2_radio_btn3.pack(padx=1, side='left')
SD3_P2_radio_btn4 = tk.Radiobutton(SD3_P2_frame, text='4', variable=SD3_P2_val, value='4')    # 放入第一個單選按鈕
SD3_P2_radio_btn4.pack(padx=1, side='left')
SD3_P2_radio_btn5 = tk.Radiobutton(SD3_P2_frame, text='5', variable=SD3_P2_val, value='5')    # 放入第一個單選按鈕
SD3_P2_radio_btn5.pack(padx=1, side='left')
#---
SD3_P3_frame = tk.Frame(SD3_P_frame, pady=5)                  # 加入 Frame 框架
SD3_P3_frame.pack(fill = 'x')
SD3_P3_val = tk.StringVar()
SD3_P3_val.set('3')
SD3_P3_a = tk.StringVar()  # 設定 a 為文字變數
SD3_P3_a.set("3. Payback needs to be quick and nasty.")    # 設定 a 的內容
SD3_P3_label = tk.Label(SD3_P3_frame, textvariable=SD3_P3_a, font=('Consolas',10))  # 放入標籤
SD3_P3_label.pack(side='left')
SD3_P3_label2 = tk.Label(SD3_P3_frame, textvariable=SD3_P3_val, font=('Consolas',10))  # 放入標籤
SD3_P3_label2.pack(side='left')
SD3_P3_radio_btn1 = tk.Radiobutton(SD3_P3_frame, text='1', variable=SD3_P3_val, value='1')    # 放入第一個單選按鈕
SD3_P3_radio_btn1.pack(padx=1, side='left')
SD3_P3_radio_btn2 = tk.Radiobutton(SD3_P3_frame, text='2', variable=SD3_P3_val, value='2')    # 放入第一個單選按鈕
SD3_P3_radio_btn2.pack(padx=1, side='left')
SD3_P3_radio_btn3 = tk.Radiobutton(SD3_P3_frame, text='3', variable=SD3_P3_val, value='3')    # 放入第一個單選按鈕
SD3_P3_radio_btn3.pack(padx=1, side='left')
SD3_P3_radio_btn4 = tk.Radiobutton(SD3_P3_frame, text='4', variable=SD3_P3_val, value='4')    # 放入第一個單選按鈕
SD3_P3_radio_btn4.pack(padx=1, side='left')
SD3_P3_radio_btn5 = tk.Radiobutton(SD3_P3_frame, text='5', variable=SD3_P3_val, value='5')    # 放入第一個單選按鈕
SD3_P3_radio_btn5.pack(padx=1, side='left')
#---
SD3_P4_frame = tk.Frame(SD3_P_frame, pady=5)                  # 加入 Frame 框架
SD3_P4_frame.pack(fill = 'x')
SD3_P4_val = tk.StringVar()
SD3_P4_val.set('3')
SD3_P4_a = tk.StringVar()  # 設定 a 為文字變數
SD3_P4_a.set("4. People often say I’m out of control.")    # 設定 a 的內容
SD3_P4_label = tk.Label(SD3_P4_frame, textvariable=SD3_P4_a, font=('Consolas',10))  # 放入標籤
SD3_P4_label.pack(side='left')
SD3_P4_label2 = tk.Label(SD3_P4_frame, textvariable=SD3_P4_val, font=('Consolas',10))  # 放入標籤
SD3_P4_label2.pack(side='left')
SD3_P4_radio_btn1 = tk.Radiobutton(SD3_P2_frame, text='1', variable=SD3_P4_val, value='1')    # 放入第一個單選按鈕
SD3_P4_radio_btn1.pack(padx=1, side='left')
SD3_P4_radio_btn2 = tk.Radiobutton(SD3_P4_frame, text='2', variable=SD3_P4_val, value='2')    # 放入第一個單選按鈕
SD3_P4_radio_btn2.pack(padx=1, side='left')
SD3_P4_radio_btn3 = tk.Radiobutton(SD3_P4_frame, text='3', variable=SD3_P4_val, value='3')    # 放入第一個單選按鈕
SD3_P4_radio_btn3.pack(padx=1, side='left')
SD3_P4_radio_btn4 = tk.Radiobutton(SD3_P4_frame, text='4', variable=SD3_P4_val, value='4')    # 放入第一個單選按鈕
SD3_P4_radio_btn4.pack(padx=1, side='left')
SD3_P4_radio_btn5 = tk.Radiobutton(SD3_P4_frame, text='5', variable=SD3_P4_val, value='5')    # 放入第一個單選按鈕
SD3_P4_radio_btn5.pack(padx=1, side='left')
#---
SD3_P5_frame = tk.Frame(SD3_P_frame, pady=5)                  # 加入 Frame 框架
SD3_P5_frame.pack(fill = 'x')
SD3_P5_val = tk.StringVar()
SD3_P5_val.set('3')
SD3_P5_a = tk.StringVar()  # 設定 a 為文字變數
SD3_P5_a.set("5. It’s true that I can be mean to others.")    # 設定 a 的內容
SD3_P5_label = tk.Label(SD3_P5_frame, textvariable=SD3_P5_a, font=('Consolas',10))  # 放入標籤
SD3_P5_label.pack(side='left')
SD3_P5_label2 = tk.Label(SD3_P5_frame, textvariable=SD3_P5_val, font=('Consolas',10))  # 放入標籤
SD3_P5_label2.pack(side='left')
SD3_P5_radio_btn1 = tk.Radiobutton(SD3_P5_frame, text='1', variable=SD3_P5_val, value='1')    # 放入第一個單選按鈕
SD3_P5_radio_btn1.pack(padx=1, side='left')
SD3_P5_radio_btn2 = tk.Radiobutton(SD3_P5_frame, text='2', variable=SD3_P5_val, value='2')    # 放入第一個單選按鈕
SD3_P5_radio_btn2.pack(padx=1, side='left')
SD3_P5_radio_btn3 = tk.Radiobutton(SD3_P5_frame, text='3', variable=SD3_P5_val, value='3')    # 放入第一個單選按鈕
SD3_P5_radio_btn3.pack(padx=1, side='left')
SD3_P5_radio_btn4 = tk.Radiobutton(SD3_P5_frame, text='4', variable=SD3_P5_val, value='4')    # 放入第一個單選按鈕
SD3_P5_radio_btn4.pack(padx=1, side='left')
SD3_P5_radio_btn5 = tk.Radiobutton(SD3_P5_frame, text='5', variable=SD3_P5_val, value='5')    # 放入第一個單選按鈕
SD3_P5_radio_btn5.pack(padx=1, side='left')
#---
SD3_P6_frame = tk.Frame(SD3_P_frame, pady=5)                  # 加入 Frame 框架
SD3_P6_frame.pack(fill = 'x')
SD3_P6_val = tk.StringVar()
SD3_P6_val.set('3')
SD3_P6_a = tk.StringVar()  # 設定 a 為文字變數
SD3_P6_a.set("6.  People who mess with me always regret it.")    # 設定 a 的內容
SD3_P6_label = tk.Label(SD3_P6_frame, textvariable=SD3_P6_a, font=('Consolas',10))  # 放入標籤
SD3_P6_label.pack(side='left')
SD3_P6_label2 = tk.Label(SD3_P6_frame, textvariable=SD3_P6_val, font=('Consolas',10))  # 放入標籤
SD3_P6_label2.pack(side='left')
SD3_P6_radio_btn1 = tk.Radiobutton(SD3_P6_frame, text='1', variable=SD3_P6_val, value='1')    # 放入第一個單選按鈕
SD3_P6_radio_btn1.pack(padx=1, side='left')
SD3_P6_radio_btn2 = tk.Radiobutton(SD3_P6_frame, text='2', variable=SD3_P6_val, value='2')    # 放入第一個單選按鈕
SD3_P6_radio_btn2.pack(padx=1, side='left')
SD3_P6_radio_btn3 = tk.Radiobutton(SD3_P6_frame, text='3', variable=SD3_P6_val, value='3')    # 放入第一個單選按鈕
SD3_P6_radio_btn3.pack(padx=1, side='left')
SD3_P6_radio_btn4 = tk.Radiobutton(SD3_P6_frame, text='4', variable=SD3_P6_val, value='4')    # 放入第一個單選按鈕
SD3_P6_radio_btn4.pack(padx=1, side='left')
SD3_P6_radio_btn5 = tk.Radiobutton(SD3_P6_frame, text='5', variable=SD3_P6_val, value='5')    # 放入第一個單選按鈕
SD3_P6_radio_btn5.pack(padx=1, side='left')
#---
SD3_P7_frame = tk.Frame(SD3_P_frame, pady=5)                  # 加入 Frame 框架
SD3_P7_frame.pack(fill = 'x')
SD3_P7_val = tk.StringVar()
SD3_P7_val.set('3')
SD3_P7_a = tk.StringVar()  # 設定 a 為文字變數
SD3_P7_a.set("7.  I have never gotten into trouble with the law. (R)")    # 設定 a 的內容
SD3_P7_label = tk.Label(SD3_P7_frame, textvariable=SD3_P7_a, font=('Consolas',10))  # 放入標籤
SD3_P7_label.pack(side='left')
SD3_P7_label2 = tk.Label(SD3_P7_frame, textvariable=SD3_P7_val, font=('Consolas',10))  # 放入標籤
SD3_P7_label2.pack(side='left')
SD3_P7_radio_btn1 = tk.Radiobutton(SD3_P7_frame, text='1', variable=SD3_P7_val, value='1')    # 放入第一個單選按鈕
SD3_P7_radio_btn1.pack(padx=1, side='left')
SD3_P7_radio_btn2 = tk.Radiobutton(SD3_P7_frame, text='2', variable=SD3_P7_val, value='2')    # 放入第一個單選按鈕
SD3_P7_radio_btn2.pack(padx=1, side='left')
SD3_P7_radio_btn3 = tk.Radiobutton(SD3_P7_frame, text='3', variable=SD3_P7_val, value='3')    # 放入第一個單選按鈕
SD3_P7_radio_btn3.pack(padx=1, side='left')
SD3_P7_radio_btn4 = tk.Radiobutton(SD3_P7_frame, text='4', variable=SD3_P7_val, value='4')    # 放入第一個單選按鈕
SD3_P7_radio_btn4.pack(padx=1, side='left')
SD3_P7_radio_btn5 = tk.Radiobutton(SD3_P7_frame, text='5', variable=SD3_P7_val, value='5')    # 放入第一個單選按鈕
SD3_P7_radio_btn5.pack(padx=1, side='left')
#---
SD3_P8_frame = tk.Frame(SD3_P_frame, pady=5)                  # 加入 Frame 框架
SD3_P8_frame.pack(fill = 'x')
SD3_P8_val = tk.StringVar()
SD3_P8_val.set('3')
SD3_P8_a = tk.StringVar()  # 設定 a 為文字變數
SD3_P8_a.set("8.  I enjoy having sex with people I hardly know")    # 設定 a 的內容
SD3_P8_label = tk.Label(SD3_P8_frame, textvariable=SD3_P8_a, font=('Consolas',10))  # 放入標籤
SD3_P8_label.pack(side='left')
SD3_P8_label2 = tk.Label(SD3_P8_frame, textvariable=SD3_P8_val, font=('Consolas',10))  # 放入標籤
SD3_P8_label2.pack(side='left')
SD3_P8_radio_btn1 = tk.Radiobutton(SD3_P8_frame, text='1', variable=SD3_P8_val, value='1')    # 放入第一個單選按鈕
SD3_P8_radio_btn1.pack(padx=1, side='left')
SD3_P8_radio_btn2 = tk.Radiobutton(SD3_P8_frame, text='2', variable=SD3_P8_val, value='2')    # 放入第一個單選按鈕
SD3_P8_radio_btn2.pack(padx=1, side='left')
SD3_P8_radio_btn3 = tk.Radiobutton(SD3_P8_frame, text='3', variable=SD3_P8_val, value='3')    # 放入第一個單選按鈕
SD3_P8_radio_btn3.pack(padx=1, side='left')
SD3_P8_radio_btn4 = tk.Radiobutton(SD3_P8_frame, text='4', variable=SD3_P8_val, value='4')    # 放入第一個單選按鈕
SD3_P8_radio_btn4.pack(padx=1, side='left')
SD3_P8_radio_btn5 = tk.Radiobutton(SD3_P8_frame, text='5', variable=SD3_P8_val, value='5')    # 放入第一個單選按鈕
SD3_P8_radio_btn5.pack(padx=1, side='left')
#---
SD3_P9_frame = tk.Frame(SD3_P_frame, pady=5)                  # 加入 Frame 框架
SD3_P9_frame.pack(fill = 'x')
SD3_P9_val = tk.StringVar()
SD3_P9_val.set('3')
SD3_P9_a = tk.StringVar()  # 設定 a 為文字變數
SD3_P9_a.set("9.  I’ll say anything to get what I want.")    # 設定 a 的內容
SD3_P9_label = tk.Label(SD3_P9_frame, textvariable=SD3_P9_a, font=('Consolas',10))  # 放入標籤
SD3_P9_label.pack(side='left')
SD3_P9_label2 = tk.Label(SD3_P9_frame, textvariable=SD3_P9_val, font=('Consolas',10))  # 放入標籤
SD3_P9_label2.pack(side='left')
SD3_P9_radio_btn1 = tk.Radiobutton(SD3_P9_frame, text='1', variable=SD3_P9_val, value='1')    # 放入第一個單選按鈕
SD3_P9_radio_btn1.pack(padx=1, side='left')
SD3_P9_radio_btn2 = tk.Radiobutton(SD3_P9_frame, text='2', variable=SD3_P9_val, value='2')    # 放入第一個單選按鈕
SD3_P9_radio_btn2.pack(padx=1, side='left')
SD3_P9_radio_btn3 = tk.Radiobutton(SD3_P9_frame, text='3', variable=SD3_P9_val, value='3')    # 放入第一個單選按鈕
SD3_P9_radio_btn3.pack(padx=1, side='left')
SD3_P9_radio_btn4 = tk.Radiobutton(SD3_P9_frame, text='4', variable=SD3_P9_val, value='4')    # 放入第一個單選按鈕
SD3_P9_radio_btn4.pack(padx=1, side='left')
SD3_P9_radio_btn5 = tk.Radiobutton(SD3_P9_frame, text='5', variable=SD3_P9_val, value='5')    # 放入第一個單選按鈕
SD3_P9_radio_btn5.pack(padx=1, side='left')
# ------------------------------------------------------------------------
IPIP_frame = tk.Frame(total_frame.interior, pady=5, padx=5)                  # 加入 Frame 框架(P )
IPIP_frame.pack(fill = 'both')
canvas = tk.Canvas(IPIP_frame, width=1200, height=50)
canvas.create_text(500, 20, text='IPIP Measures Questionnaire Extraversion', anchor='center', font=('Consolas', 14))
canvas.create_text(500, 40, text='Very Inaccurate = 1, Moderately Inaccurate = 2, Neither Accurate Nor Inaccurate = 3, Moderately Accurate = 4, Very Accurate = 5', anchor='center', font=('Consolas', 10))
canvas.pack(pady=5)
# ------------------------------------------------------------------------
IPIP_01_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_01_frame.pack(fill = 'x')
IPIP_01_val = tk.StringVar()
IPIP_01_val.set('3')
IPIP_01_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_01_a.set("1. Am the life of the party.")    # 設定 a 的內容
IPIP_01_label = tk.Label(IPIP_01_frame, textvariable=IPIP_01_a, font=('Consolas',10))  # 放入標籤
IPIP_01_label.pack(side='left')
IPIP_01_label2 = tk.Label(IPIP_01_frame, textvariable=IPIP_01_val, font=('Consolas',10))  # 放入標籤
IPIP_01_label2.pack(side='left')
IPIP_01_radio_btn1 = tk.Radiobutton(IPIP_01_frame, text='1', variable=IPIP_01_val, value='1')    # 放入第一個單選按鈕
IPIP_01_radio_btn1.pack(padx=1, side='left')
IPIP_01_radio_btn2 = tk.Radiobutton(IPIP_01_frame, text='2', variable=IPIP_01_val, value='2')    # 放入第一個單選按鈕
IPIP_01_radio_btn2.pack(padx=1, side='left')
IPIP_01_radio_btn3 = tk.Radiobutton(IPIP_01_frame, text='3', variable=IPIP_01_val, value='3')    # 放入第一個單選按鈕
IPIP_01_radio_btn3.pack(padx=1, side='left')
IPIP_01_radio_btn4 = tk.Radiobutton(IPIP_01_frame, text='4', variable=IPIP_01_val, value='4')    # 放入第一個單選按鈕
IPIP_01_radio_btn4.pack(padx=1, side='left')
IPIP_01_radio_btn5 = tk.Radiobutton(IPIP_01_frame, text='5', variable=IPIP_01_val, value='5')    # 放入第一個單選按鈕
IPIP_01_radio_btn5.pack(padx=1, side='left')
#---
IPIP_02_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_02_frame.pack(fill = 'x')
IPIP_02_val = tk.StringVar()
IPIP_02_val.set('3')
IPIP_02_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_02_a.set("2. Feel little concern for others.")    # 設定 a 的內容
IPIP_02_label = tk.Label(IPIP_02_frame, textvariable=IPIP_02_a, font=('Consolas',10))  # 放入標籤
IPIP_02_label.pack(side='left')
IPIP_02_label2 = tk.Label(IPIP_02_frame, textvariable=IPIP_02_val, font=('Consolas',10))  # 放入標籤
IPIP_02_label2.pack(side='left')
IPIP_02_radio_btn1 = tk.Radiobutton(IPIP_02_frame, text='1', variable=IPIP_02_val, value='1')    # 放入第一個單選按鈕
IPIP_02_radio_btn1.pack(padx=1, side='left')
IPIP_02_radio_btn2 = tk.Radiobutton(IPIP_02_frame, text='2', variable=IPIP_02_val, value='2')    # 放入第一個單選按鈕
IPIP_02_radio_btn2.pack(padx=1, side='left')
IPIP_02_radio_btn3 = tk.Radiobutton(IPIP_02_frame, text='3', variable=IPIP_02_val, value='3')    # 放入第一個單選按鈕
IPIP_02_radio_btn3.pack(padx=1, side='left')
IPIP_02_radio_btn4 = tk.Radiobutton(IPIP_02_frame, text='4', variable=IPIP_02_val, value='4')    # 放入第一個單選按鈕
IPIP_02_radio_btn4.pack(padx=1, side='left')
IPIP_02_radio_btn5 = tk.Radiobutton(IPIP_02_frame, text='5', variable=IPIP_02_val, value='5')    # 放入第一個單選按鈕
IPIP_02_radio_btn5.pack(padx=1, side='left')
#---
IPIP_03_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_03_frame.pack(fill = 'x')
IPIP_03_val = tk.StringVar()
IPIP_03_val.set('3')
IPIP_03_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_03_a.set("3. Am always prepared.")    # 設定 a 的內容
IPIP_03_label = tk.Label(IPIP_03_frame, textvariable=IPIP_03_a, font=('Consolas',10))  # 放入標籤
IPIP_03_label.pack(side='left')
IPIP_03_label2 = tk.Label(IPIP_03_frame, textvariable=IPIP_03_val, font=('Consolas',10))  # 放入標籤
IPIP_03_label2.pack(side='left')
IPIP_03_radio_btn1 = tk.Radiobutton(IPIP_03_frame, text='1', variable=IPIP_03_val, value='1')    # 放入第一個單選按鈕
IPIP_03_radio_btn1.pack(padx=1, side='left')
IPIP_03_radio_btn2 = tk.Radiobutton(IPIP_03_frame, text='2', variable=IPIP_03_val, value='2')    # 放入第一個單選按鈕
IPIP_03_radio_btn2.pack(padx=1, side='left')
IPIP_03_radio_btn3 = tk.Radiobutton(IPIP_03_frame, text='3', variable=IPIP_03_val, value='3')    # 放入第一個單選按鈕
IPIP_03_radio_btn3.pack(padx=1, side='left')
IPIP_03_radio_btn4 = tk.Radiobutton(IPIP_03_frame, text='4', variable=IPIP_03_val, value='4')    # 放入第一個單選按鈕
IPIP_03_radio_btn4.pack(padx=1, side='left')
IPIP_03_radio_btn5 = tk.Radiobutton(IPIP_03_frame, text='5', variable=IPIP_03_val, value='5')    # 放入第一個單選按鈕
IPIP_03_radio_btn5.pack(padx=1, side='left')
#---
IPIP_04_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_04_frame.pack(fill = 'x')
IPIP_04_val = tk.StringVar()
IPIP_04_val.set('3')
IPIP_04_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_04_a.set("4. Get stressed out easily.")    # 設定 a 的內容
IPIP_04_label = tk.Label(IPIP_04_frame, textvariable=IPIP_04_a, font=('Consolas',10))  # 放入標籤
IPIP_04_label.pack(side='left')
IPIP_04_label2 = tk.Label(IPIP_04_frame, textvariable=IPIP_04_val, font=('Consolas',10))  # 放入標籤
IPIP_04_label2.pack(side='left')
IPIP_04_radio_btn1 = tk.Radiobutton(IPIP_04_frame, text='1', variable=IPIP_04_val, value='1')    # 放入第一個單選按鈕
IPIP_04_radio_btn1.pack(padx=1, side='left')
IPIP_04_radio_btn2 = tk.Radiobutton(IPIP_04_frame, text='2', variable=IPIP_04_val, value='2')    # 放入第一個單選按鈕
IPIP_04_radio_btn2.pack(padx=1, side='left')
IPIP_04_radio_btn3 = tk.Radiobutton(IPIP_04_frame, text='3', variable=IPIP_04_val, value='3')    # 放入第一個單選按鈕
IPIP_04_radio_btn3.pack(padx=1, side='left')
IPIP_04_radio_btn4 = tk.Radiobutton(IPIP_04_frame, text='4', variable=IPIP_04_val, value='4')    # 放入第一個單選按鈕
IPIP_04_radio_btn4.pack(padx=1, side='left')
IPIP_04_radio_btn5 = tk.Radiobutton(IPIP_04_frame, text='5', variable=IPIP_04_val, value='5')    # 放入第一個單選按鈕
IPIP_04_radio_btn5.pack(padx=1, side='left')
#---
IPIP_05_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_05_frame.pack(fill = 'x')
IPIP_05_val = tk.StringVar()
IPIP_05_val.set('3')
IPIP_05_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_05_a.set("5. Have a rich vocabulary.")    # 設定 a 的內容
IPIP_05_label = tk.Label(IPIP_05_frame, textvariable=IPIP_05_a, font=('Consolas',10))  # 放入標籤
IPIP_05_label.pack(side='left')
IPIP_05_label2 = tk.Label(IPIP_05_frame, textvariable=IPIP_05_val, font=('Consolas',10))  # 放入標籤
IPIP_05_label2.pack(side='left')
IPIP_05_radio_btn1 = tk.Radiobutton(IPIP_05_frame, text='1', variable=IPIP_05_val, value='1')    # 放入第一個單選按鈕
IPIP_05_radio_btn1.pack(padx=1, side='left')
IPIP_05_radio_btn2 = tk.Radiobutton(IPIP_05_frame, text='2', variable=IPIP_05_val, value='2')    # 放入第一個單選按鈕
IPIP_05_radio_btn2.pack(padx=1, side='left')
IPIP_05_radio_btn3 = tk.Radiobutton(IPIP_05_frame, text='3', variable=IPIP_05_val, value='3')    # 放入第一個單選按鈕
IPIP_05_radio_btn3.pack(padx=1, side='left')
IPIP_05_radio_btn4 = tk.Radiobutton(IPIP_05_frame, text='4', variable=IPIP_05_val, value='4')    # 放入第一個單選按鈕
IPIP_05_radio_btn4.pack(padx=1, side='left')
IPIP_05_radio_btn5 = tk.Radiobutton(IPIP_05_frame, text='5', variable=IPIP_05_val, value='5')    # 放入第一個單選按鈕
IPIP_05_radio_btn5.pack(padx=1, side='left')
#---
IPIP_06_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_06_frame.pack(fill = 'x')
IPIP_06_val = tk.StringVar()
IPIP_06_val.set('3')
IPIP_06_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_06_a.set("6. Don't talk a lot.")    # 設定 a 的內容
IPIP_06_label = tk.Label(IPIP_06_frame, textvariable=IPIP_06_a, font=('Consolas',10))  # 放入標籤
IPIP_06_label.pack(side='left')
IPIP_06_label2 = tk.Label(IPIP_06_frame, textvariable=IPIP_06_val, font=('Consolas',10))  # 放入標籤
IPIP_06_label2.pack(side='left')
IPIP_06_radio_btn1 = tk.Radiobutton(IPIP_06_frame, text='1', variable=IPIP_06_val, value='1')    # 放入第一個單選按鈕
IPIP_06_radio_btn1.pack(padx=1, side='left')
IPIP_06_radio_btn2 = tk.Radiobutton(IPIP_06_frame, text='2', variable=IPIP_06_val, value='2')    # 放入第一個單選按鈕
IPIP_06_radio_btn2.pack(padx=1, side='left')
IPIP_06_radio_btn3 = tk.Radiobutton(IPIP_06_frame, text='3', variable=IPIP_06_val, value='3')    # 放入第一個單選按鈕
IPIP_06_radio_btn3.pack(padx=1, side='left')
IPIP_06_radio_btn4 = tk.Radiobutton(IPIP_06_frame, text='4', variable=IPIP_06_val, value='4')    # 放入第一個單選按鈕
IPIP_06_radio_btn4.pack(padx=1, side='left')
IPIP_06_radio_btn5 = tk.Radiobutton(IPIP_06_frame, text='5', variable=IPIP_06_val, value='5')    # 放入第一個單選按鈕
IPIP_06_radio_btn5.pack(padx=1, side='left')
#---
IPIP_07_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_07_frame.pack(fill = 'x')
IPIP_07_val = tk.StringVar()
IPIP_07_val.set('3')
IPIP_07_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_07_a.set("7. Am interested in people.")    # 設定 a 的內容
IPIP_07_label = tk.Label(IPIP_07_frame, textvariable=IPIP_07_a, font=('Consolas',10))  # 放入標籤
IPIP_07_label.pack(side='left')
IPIP_07_label2 = tk.Label(IPIP_07_frame, textvariable=IPIP_07_val, font=('Consolas',10))  # 放入標籤
IPIP_07_label2.pack(side='left')
IPIP_07_radio_btn1 = tk.Radiobutton(IPIP_07_frame, text='1', variable=IPIP_07_val, value='1')    # 放入第一個單選按鈕
IPIP_07_radio_btn1.pack(padx=1, side='left')
IPIP_07_radio_btn2 = tk.Radiobutton(IPIP_07_frame, text='2', variable=IPIP_07_val, value='2')    # 放入第一個單選按鈕
IPIP_07_radio_btn2.pack(padx=1, side='left')
IPIP_07_radio_btn3 = tk.Radiobutton(IPIP_07_frame, text='3', variable=IPIP_07_val, value='3')    # 放入第一個單選按鈕
IPIP_07_radio_btn3.pack(padx=1, side='left')
IPIP_07_radio_btn4 = tk.Radiobutton(IPIP_07_frame, text='4', variable=IPIP_07_val, value='4')    # 放入第一個單選按鈕
IPIP_07_radio_btn4.pack(padx=1, side='left')
IPIP_07_radio_btn5 = tk.Radiobutton(IPIP_07_frame, text='5', variable=IPIP_07_val, value='5')    # 放入第一個單選按鈕
IPIP_07_radio_btn5.pack(padx=1, side='left')
#---
IPIP_08_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_08_frame.pack(fill = 'x')
IPIP_08_val = tk.StringVar()
IPIP_08_val.set('3')
IPIP_08_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_08_a.set("8. Leave my belongings around.")    # 設定 a 的內容
IPIP_08_label = tk.Label(IPIP_08_frame, textvariable=IPIP_08_a, font=('Consolas',10))  # 放入標籤
IPIP_08_label.pack(side='left')
IPIP_08_label2 = tk.Label(IPIP_08_frame, textvariable=IPIP_08_val, font=('Consolas',10))  # 放入標籤
IPIP_08_label2.pack(side='left')
IPIP_08_radio_btn1 = tk.Radiobutton(IPIP_08_frame, text='1', variable=IPIP_08_val, value='1')    # 放入第一個單選按鈕
IPIP_08_radio_btn1.pack(padx=1, side='left')
IPIP_08_radio_btn2 = tk.Radiobutton(IPIP_08_frame, text='2', variable=IPIP_08_val, value='2')    # 放入第一個單選按鈕
IPIP_08_radio_btn2.pack(padx=1, side='left')
IPIP_08_radio_btn3 = tk.Radiobutton(IPIP_08_frame, text='3', variable=IPIP_08_val, value='3')    # 放入第一個單選按鈕
IPIP_08_radio_btn3.pack(padx=1, side='left')
IPIP_08_radio_btn4 = tk.Radiobutton(IPIP_08_frame, text='4', variable=IPIP_08_val, value='4')    # 放入第一個單選按鈕
IPIP_08_radio_btn4.pack(padx=1, side='left')
IPIP_08_radio_btn5 = tk.Radiobutton(IPIP_08_frame, text='5', variable=IPIP_08_val, value='5')    # 放入第一個單選按鈕
IPIP_08_radio_btn5.pack(padx=1, side='left')
#---
IPIP_09_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_09_frame.pack(fill = 'x')
IPIP_09_val = tk.StringVar()
IPIP_09_val.set('3')
IPIP_09_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_09_a.set("9. Am relaxed most of the time.")    # 設定 a 的內容
IPIP_09_label = tk.Label(IPIP_09_frame, textvariable=IPIP_09_a, font=('Consolas',10))  # 放入標籤
IPIP_09_label.pack(side='left')
IPIP_09_label2 = tk.Label(IPIP_09_frame, textvariable=IPIP_09_val, font=('Consolas',10))  # 放入標籤
IPIP_09_label2.pack(side='left')
IPIP_09_radio_btn1 = tk.Radiobutton(IPIP_09_frame, text='1', variable=IPIP_09_val, value='1')    # 放入第一個單選按鈕
IPIP_09_radio_btn1.pack(padx=1, side='left')
IPIP_09_radio_btn2 = tk.Radiobutton(IPIP_09_frame, text='2', variable=IPIP_09_val, value='2')    # 放入第一個單選按鈕
IPIP_09_radio_btn2.pack(padx=1, side='left')
IPIP_09_radio_btn3 = tk.Radiobutton(IPIP_09_frame, text='3', variable=IPIP_09_val, value='3')    # 放入第一個單選按鈕
IPIP_09_radio_btn3.pack(padx=1, side='left')
IPIP_09_radio_btn4 = tk.Radiobutton(IPIP_09_frame, text='4', variable=IPIP_09_val, value='4')    # 放入第一個單選按鈕
IPIP_09_radio_btn4.pack(padx=1, side='left')
IPIP_09_radio_btn5 = tk.Radiobutton(IPIP_09_frame, text='5', variable=IPIP_09_val, value='5')    # 放入第一個單選按鈕
IPIP_09_radio_btn5.pack(padx=1, side='left')
#---
IPIP_10_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_10_frame.pack(fill = 'x')
IPIP_10_val = tk.StringVar()
IPIP_10_val.set('3')
IPIP_10_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_10_a.set("10.Have difficulty understanding abstract ideas.")    # 設定 a 的內容
IPIP_10_label = tk.Label(IPIP_10_frame, textvariable=IPIP_10_a, font=('Consolas',10))  # 放入標籤
IPIP_10_label.pack(side='left')
IPIP_10_label2 = tk.Label(IPIP_10_frame, textvariable=IPIP_10_val, font=('Consolas',10))  # 放入標籤
IPIP_10_label2.pack(side='left')
IPIP_10_radio_btn1 = tk.Radiobutton(IPIP_10_frame, text='1', variable=IPIP_10_val, value='1')    # 放入第一個單選按鈕
IPIP_10_radio_btn1.pack(padx=1, side='left')
IPIP_10_radio_btn2 = tk.Radiobutton(IPIP_10_frame, text='2', variable=IPIP_10_val, value='2')    # 放入第一個單選按鈕
IPIP_10_radio_btn2.pack(padx=1, side='left')
IPIP_10_radio_btn3 = tk.Radiobutton(IPIP_10_frame, text='3', variable=IPIP_10_val, value='3')    # 放入第一個單選按鈕
IPIP_10_radio_btn3.pack(padx=1, side='left')
IPIP_10_radio_btn4 = tk.Radiobutton(IPIP_10_frame, text='4', variable=IPIP_10_val, value='4')    # 放入第一個單選按鈕
IPIP_10_radio_btn4.pack(padx=1, side='left')
IPIP_10_radio_btn5 = tk.Radiobutton(IPIP_10_frame, text='5', variable=IPIP_10_val, value='5')    # 放入第一個單選按鈕
IPIP_10_radio_btn5.pack(padx=1, side='left')
#---
IPIP_11_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_11_frame.pack(fill = 'x')
IPIP_11_val = tk.StringVar()
IPIP_11_val.set('3')
IPIP_11_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_11_a.set("11.Feel comfortable around people.")    # 設定 a 的內容
IPIP_11_label = tk.Label(IPIP_11_frame, textvariable=IPIP_11_a, font=('Consolas',10))  # 放入標籤
IPIP_11_label.pack(side='left')
IPIP_11_label2 = tk.Label(IPIP_11_frame, textvariable=IPIP_11_val, font=('Consolas',10))  # 放入標籤
IPIP_11_label2.pack(side='left')
IPIP_11_radio_btn1 = tk.Radiobutton(IPIP_11_frame, text='1', variable=IPIP_11_val, value='1')    # 放入第一個單選按鈕
IPIP_11_radio_btn1.pack(padx=1, side='left')
IPIP_11_radio_btn2 = tk.Radiobutton(IPIP_11_frame, text='2', variable=IPIP_11_val, value='2')    # 放入第一個單選按鈕
IPIP_11_radio_btn2.pack(padx=1, side='left')
IPIP_11_radio_btn3 = tk.Radiobutton(IPIP_11_frame, text='3', variable=IPIP_11_val, value='3')    # 放入第一個單選按鈕
IPIP_11_radio_btn3.pack(padx=1, side='left')
IPIP_11_radio_btn4 = tk.Radiobutton(IPIP_11_frame, text='4', variable=IPIP_11_val, value='4')    # 放入第一個單選按鈕
IPIP_11_radio_btn4.pack(padx=1, side='left')
IPIP_11_radio_btn5 = tk.Radiobutton(IPIP_11_frame, text='5', variable=IPIP_11_val, value='5')    # 放入第一個單選按鈕
IPIP_11_radio_btn5.pack(padx=1, side='left')
#---
IPIP_12_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_12_frame.pack(fill = 'x')
IPIP_12_val = tk.StringVar()
IPIP_12_val.set('3')
IPIP_12_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_12_a.set("12.Insult people.")    # 設定 a 的內容
IPIP_12_label = tk.Label(IPIP_12_frame, textvariable=IPIP_12_a, font=('Consolas',10))  # 放入標籤
IPIP_12_label.pack(side='left')
IPIP_12_label2 = tk.Label(IPIP_12_frame, textvariable=IPIP_12_val, font=('Consolas',10))  # 放入標籤
IPIP_12_label2.pack(side='left')
IPIP_12_radio_btn1 = tk.Radiobutton(IPIP_12_frame, text='1', variable=IPIP_12_val, value='1')    # 放入第一個單選按鈕
IPIP_12_radio_btn1.pack(padx=1, side='left')
IPIP_12_radio_btn2 = tk.Radiobutton(IPIP_12_frame, text='2', variable=IPIP_12_val, value='2')    # 放入第一個單選按鈕
IPIP_12_radio_btn2.pack(padx=1, side='left')
IPIP_12_radio_btn3 = tk.Radiobutton(IPIP_12_frame, text='3', variable=IPIP_12_val, value='3')    # 放入第一個單選按鈕
IPIP_12_radio_btn3.pack(padx=1, side='left')
IPIP_12_radio_btn4 = tk.Radiobutton(IPIP_12_frame, text='4', variable=IPIP_12_val, value='4')    # 放入第一個單選按鈕
IPIP_12_radio_btn4.pack(padx=1, side='left')
IPIP_12_radio_btn5 = tk.Radiobutton(IPIP_12_frame, text='5', variable=IPIP_12_val, value='5')    # 放入第一個單選按鈕
IPIP_12_radio_btn5.pack(padx=1, side='left')
#---
IPIP_13_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_13_frame.pack(fill = 'x')
IPIP_13_val = tk.StringVar()
IPIP_13_val.set('3')
IPIP_13_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_13_a.set("13.Pay attention to details.")    # 設定 a 的內容
IPIP_13_label = tk.Label(IPIP_13_frame, textvariable=IPIP_13_a, font=('Consolas',10))  # 放入標籤
IPIP_13_label.pack(side='left')
IPIP_13_label2 = tk.Label(IPIP_13_frame, textvariable=IPIP_13_val, font=('Consolas',10))  # 放入標籤
IPIP_13_label2.pack(side='left')
IPIP_13_radio_btn1 = tk.Radiobutton(IPIP_13_frame, text='1', variable=IPIP_13_val, value='1')    # 放入第一個單選按鈕
IPIP_13_radio_btn1.pack(padx=1, side='left')
IPIP_13_radio_btn2 = tk.Radiobutton(IPIP_13_frame, text='2', variable=IPIP_13_val, value='2')    # 放入第一個單選按鈕
IPIP_13_radio_btn2.pack(padx=1, side='left')
IPIP_13_radio_btn3 = tk.Radiobutton(IPIP_13_frame, text='3', variable=IPIP_13_val, value='3')    # 放入第一個單選按鈕
IPIP_13_radio_btn3.pack(padx=1, side='left')
IPIP_13_radio_btn4 = tk.Radiobutton(IPIP_13_frame, text='4', variable=IPIP_13_val, value='4')    # 放入第一個單選按鈕
IPIP_13_radio_btn4.pack(padx=1, side='left')
IPIP_13_radio_btn5 = tk.Radiobutton(IPIP_13_frame, text='5', variable=IPIP_13_val, value='5')    # 放入第一個單選按鈕
IPIP_13_radio_btn5.pack(padx=1, side='left')
#---
IPIP_14_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_14_frame.pack(fill = 'x')
IPIP_14_val = tk.StringVar()
IPIP_14_val.set('3')
IPIP_14_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_14_a.set("14.Worry about things.")    # 設定 a 的內容
IPIP_14_label = tk.Label(IPIP_14_frame, textvariable=IPIP_14_a, font=('Consolas',10))  # 放入標籤
IPIP_14_label.pack(side='left')
IPIP_14_label2 = tk.Label(IPIP_14_frame, textvariable=IPIP_14_val, font=('Consolas',10))  # 放入標籤
IPIP_14_label2.pack(side='left')
IPIP_14_radio_btn1 = tk.Radiobutton(IPIP_14_frame, text='1', variable=IPIP_14_val, value='1')    # 放入第一個單選按鈕
IPIP_14_radio_btn1.pack(padx=1, side='left')
IPIP_14_radio_btn2 = tk.Radiobutton(IPIP_14_frame, text='2', variable=IPIP_14_val, value='2')    # 放入第一個單選按鈕
IPIP_14_radio_btn2.pack(padx=1, side='left')
IPIP_14_radio_btn3 = tk.Radiobutton(IPIP_14_frame, text='3', variable=IPIP_14_val, value='3')    # 放入第一個單選按鈕
IPIP_14_radio_btn3.pack(padx=1, side='left')
IPIP_14_radio_btn4 = tk.Radiobutton(IPIP_14_frame, text='4', variable=IPIP_14_val, value='4')    # 放入第一個單選按鈕
IPIP_14_radio_btn4.pack(padx=1, side='left')
IPIP_14_radio_btn5 = tk.Radiobutton(IPIP_14_frame, text='5', variable=IPIP_14_val, value='5')    # 放入第一個單選按鈕
IPIP_14_radio_btn5.pack(padx=1, side='left')
#---
IPIP_15_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_15_frame.pack(fill = 'x')
IPIP_15_val = tk.StringVar()
IPIP_15_val.set('3')
IPIP_15_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_15_a.set("15.Have a rich vocabulary.")    # 設定 a 的內容
IPIP_15_label = tk.Label(IPIP_15_frame, textvariable=IPIP_15_a, font=('Consolas',10))  # 放入標籤
IPIP_15_label.pack(side='left')
IPIP_15_label2 = tk.Label(IPIP_15_frame, textvariable=IPIP_15_val, font=('Consolas',10))  # 放入標籤
IPIP_15_label2.pack(side='left')
IPIP_15_radio_btn1 = tk.Radiobutton(IPIP_15_frame, text='1', variable=IPIP_15_val, value='1')    # 放入第一個單選按鈕
IPIP_15_radio_btn1.pack(padx=1, side='left')
IPIP_15_radio_btn2 = tk.Radiobutton(IPIP_15_frame, text='2', variable=IPIP_15_val, value='2')    # 放入第一個單選按鈕
IPIP_15_radio_btn2.pack(padx=1, side='left')
IPIP_15_radio_btn3 = tk.Radiobutton(IPIP_15_frame, text='3', variable=IPIP_15_val, value='3')    # 放入第一個單選按鈕
IPIP_15_radio_btn3.pack(padx=1, side='left')
IPIP_15_radio_btn4 = tk.Radiobutton(IPIP_15_frame, text='4', variable=IPIP_15_val, value='4')    # 放入第一個單選按鈕
IPIP_15_radio_btn4.pack(padx=1, side='left')
IPIP_15_radio_btn5 = tk.Radiobutton(IPIP_15_frame, text='5', variable=IPIP_15_val, value='5')    # 放入第一個單選按鈕
IPIP_15_radio_btn5.pack(padx=1, side='left')
#---
IPIP_16_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_16_frame.pack(fill = 'x')
IPIP_16_val = tk.StringVar()
IPIP_16_val.set('3')
IPIP_16_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_16_a.set("16.Don't talk a lot.")    # 設定 a 的內容
IPIP_16_label = tk.Label(IPIP_16_frame, textvariable=IPIP_16_a, font=('Consolas',10))  # 放入標籤
IPIP_16_label.pack(side='left')
IPIP_16_label2 = tk.Label(IPIP_16_frame, textvariable=IPIP_16_val, font=('Consolas',10))  # 放入標籤
IPIP_16_label2.pack(side='left')
IPIP_16_radio_btn1 = tk.Radiobutton(IPIP_16_frame, text='1', variable=IPIP_16_val, value='1')    # 放入第一個單選按鈕
IPIP_16_radio_btn1.pack(padx=1, side='left')
IPIP_16_radio_btn2 = tk.Radiobutton(IPIP_16_frame, text='2', variable=IPIP_16_val, value='2')    # 放入第一個單選按鈕
IPIP_16_radio_btn2.pack(padx=1, side='left')
IPIP_16_radio_btn3 = tk.Radiobutton(IPIP_16_frame, text='3', variable=IPIP_16_val, value='3')    # 放入第一個單選按鈕
IPIP_16_radio_btn3.pack(padx=1, side='left')
IPIP_16_radio_btn4 = tk.Radiobutton(IPIP_16_frame, text='4', variable=IPIP_16_val, value='4')    # 放入第一個單選按鈕
IPIP_16_radio_btn4.pack(padx=1, side='left')
IPIP_16_radio_btn5 = tk.Radiobutton(IPIP_16_frame, text='5', variable=IPIP_16_val, value='5')    # 放入第一個單選按鈕
IPIP_16_radio_btn5.pack(padx=1, side='left')
#---
IPIP_17_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_17_frame.pack(fill = 'x')
IPIP_17_val = tk.StringVar()
IPIP_17_val.set('3')
IPIP_17_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_17_a.set("17.Am interested in people.")    # 設定 a 的內容
IPIP_17_label = tk.Label(IPIP_17_frame, textvariable=IPIP_17_a, font=('Consolas',10))  # 放入標籤
IPIP_17_label.pack(side='left')
IPIP_17_label2 = tk.Label(IPIP_17_frame, textvariable=IPIP_17_val, font=('Consolas',10))  # 放入標籤
IPIP_17_label2.pack(side='left')
IPIP_17_radio_btn1 = tk.Radiobutton(IPIP_17_frame, text='1', variable=IPIP_17_val, value='1')    # 放入第一個單選按鈕
IPIP_17_radio_btn1.pack(padx=1, side='left')
IPIP_17_radio_btn2 = tk.Radiobutton(IPIP_17_frame, text='2', variable=IPIP_17_val, value='2')    # 放入第一個單選按鈕
IPIP_17_radio_btn2.pack(padx=1, side='left')
IPIP_17_radio_btn3 = tk.Radiobutton(IPIP_17_frame, text='3', variable=IPIP_17_val, value='3')    # 放入第一個單選按鈕
IPIP_17_radio_btn3.pack(padx=1, side='left')
IPIP_17_radio_btn4 = tk.Radiobutton(IPIP_17_frame, text='4', variable=IPIP_17_val, value='4')    # 放入第一個單選按鈕
IPIP_17_radio_btn4.pack(padx=1, side='left')
IPIP_17_radio_btn5 = tk.Radiobutton(IPIP_17_frame, text='5', variable=IPIP_17_val, value='5')    # 放入第一個單選按鈕
IPIP_17_radio_btn5.pack(padx=1, side='left')
#---
IPIP_18_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_18_frame.pack(fill = 'x')
IPIP_18_val = tk.StringVar()
IPIP_18_val.set('3')
IPIP_18_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_18_a.set("18.Leave my belongings around.")    # 設定 a 的內容
IPIP_18_label = tk.Label(IPIP_18_frame, textvariable=IPIP_18_a, font=('Consolas',10))  # 放入標籤
IPIP_18_label.pack(side='left')
IPIP_18_label2 = tk.Label(IPIP_18_frame, textvariable=IPIP_18_val, font=('Consolas',10))  # 放入標籤
IPIP_18_label2.pack(side='left')
IPIP_18_radio_btn1 = tk.Radiobutton(IPIP_18_frame, text='1', variable=IPIP_18_val, value='1')    # 放入第一個單選按鈕
IPIP_18_radio_btn1.pack(padx=1, side='left')
IPIP_18_radio_btn2 = tk.Radiobutton(IPIP_18_frame, text='2', variable=IPIP_18_val, value='2')    # 放入第一個單選按鈕
IPIP_18_radio_btn2.pack(padx=1, side='left')
IPIP_18_radio_btn3 = tk.Radiobutton(IPIP_18_frame, text='3', variable=IPIP_18_val, value='3')    # 放入第一個單選按鈕
IPIP_18_radio_btn3.pack(padx=1, side='left')
IPIP_18_radio_btn4 = tk.Radiobutton(IPIP_18_frame, text='4', variable=IPIP_18_val, value='4')    # 放入第一個單選按鈕
IPIP_18_radio_btn4.pack(padx=1, side='left')
IPIP_18_radio_btn5 = tk.Radiobutton(IPIP_18_frame, text='5', variable=IPIP_18_val, value='5')    # 放入第一個單選按鈕
IPIP_18_radio_btn5.pack(padx=1, side='left')
#---
IPIP_19_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_19_frame.pack(fill = 'x')
IPIP_19_val = tk.StringVar()
IPIP_19_val.set('3')
IPIP_19_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_19_a.set("19.Am relaxed most of the time.")    # 設定 a 的內容
IPIP_19_label = tk.Label(IPIP_19_frame, textvariable=IPIP_19_a, font=('Consolas',10))  # 放入標籤
IPIP_19_label.pack(side='left')
IPIP_19_label2 = tk.Label(IPIP_19_frame, textvariable=IPIP_19_val, font=('Consolas',10))  # 放入標籤
IPIP_19_label2.pack(side='left')
IPIP_19_radio_btn1 = tk.Radiobutton(IPIP_19_frame, text='1', variable=IPIP_19_val, value='1')    # 放入第一個單選按鈕
IPIP_19_radio_btn1.pack(padx=1, side='left')
IPIP_19_radio_btn2 = tk.Radiobutton(IPIP_19_frame, text='2', variable=IPIP_19_val, value='2')    # 放入第一個單選按鈕
IPIP_19_radio_btn2.pack(padx=1, side='left')
IPIP_19_radio_btn3 = tk.Radiobutton(IPIP_19_frame, text='3', variable=IPIP_19_val, value='3')    # 放入第一個單選按鈕
IPIP_19_radio_btn3.pack(padx=1, side='left')
IPIP_19_radio_btn4 = tk.Radiobutton(IPIP_19_frame, text='4', variable=IPIP_19_val, value='4')    # 放入第一個單選按鈕
IPIP_19_radio_btn4.pack(padx=1, side='left')
IPIP_19_radio_btn5 = tk.Radiobutton(IPIP_19_frame, text='5', variable=IPIP_19_val, value='5')    # 放入第一個單選按鈕
IPIP_19_radio_btn5.pack(padx=1, side='left')
#---
IPIP_20_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_20_frame.pack(fill = 'x')
IPIP_20_val = tk.StringVar()
IPIP_20_val.set('3')
IPIP_20_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_20_a.set("20.Have difficulty understanding abstract ideas.")    # 設定 a 的內容
IPIP_20_label = tk.Label(IPIP_20_frame, textvariable=IPIP_20_a, font=('Consolas',10))  # 放入標籤
IPIP_20_label.pack(side='left')
IPIP_20_label2 = tk.Label(IPIP_20_frame, textvariable=IPIP_20_val, font=('Consolas',10))  # 放入標籤
IPIP_20_label2.pack(side='left')
IPIP_20_radio_btn1 = tk.Radiobutton(IPIP_20_frame, text='1', variable=IPIP_20_val, value='1')    # 放入第一個單選按鈕
IPIP_20_radio_btn1.pack(padx=1, side='left')
IPIP_20_radio_btn2 = tk.Radiobutton(IPIP_20_frame, text='2', variable=IPIP_20_val, value='2')    # 放入第一個單選按鈕
IPIP_20_radio_btn2.pack(padx=1, side='left')
IPIP_20_radio_btn3 = tk.Radiobutton(IPIP_20_frame, text='3', variable=IPIP_20_val, value='3')    # 放入第一個單選按鈕
IPIP_20_radio_btn3.pack(padx=1, side='left')
IPIP_20_radio_btn4 = tk.Radiobutton(IPIP_20_frame, text='4', variable=IPIP_20_val, value='4')    # 放入第一個單選按鈕
IPIP_20_radio_btn4.pack(padx=1, side='left')
IPIP_20_radio_btn5 = tk.Radiobutton(IPIP_20_frame, text='5', variable=IPIP_20_val, value='5')    # 放入第一個單選按鈕
IPIP_20_radio_btn5.pack(padx=1, side='left')
# ------------------------------------------------------------------------
IPIP_21_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_21_frame.pack(fill = 'x')
IPIP_21_val = tk.StringVar()
IPIP_21_val.set('3')
IPIP_21_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_21_a.set("21.Start conversations.")    # 設定 a 的內容
IPIP_21_label = tk.Label(IPIP_21_frame, textvariable=IPIP_21_a, font=('Consolas',10))  # 放入標籤
IPIP_21_label.pack(side='left')
IPIP_21_label2 = tk.Label(IPIP_21_frame, textvariable=IPIP_21_val, font=('Consolas',10))  # 放入標籤
IPIP_21_label2.pack(side='left')
IPIP_21_radio_btn1 = tk.Radiobutton(IPIP_21_frame, text='1', variable=IPIP_21_val, value='1')    # 放入第一個單選按鈕
IPIP_21_radio_btn1.pack(padx=1, side='left')
IPIP_21_radio_btn2 = tk.Radiobutton(IPIP_21_frame, text='2', variable=IPIP_21_val, value='2')    # 放入第一個單選按鈕
IPIP_21_radio_btn2.pack(padx=1, side='left')
IPIP_21_radio_btn3 = tk.Radiobutton(IPIP_21_frame, text='3', variable=IPIP_21_val, value='3')    # 放入第一個單選按鈕
IPIP_21_radio_btn3.pack(padx=1, side='left')
IPIP_21_radio_btn4 = tk.Radiobutton(IPIP_21_frame, text='4', variable=IPIP_21_val, value='4')    # 放入第一個單選按鈕
IPIP_21_radio_btn4.pack(padx=1, side='left')
IPIP_21_radio_btn5 = tk.Radiobutton(IPIP_21_frame, text='5', variable=IPIP_21_val, value='5')    # 放入第一個單選按鈕
IPIP_21_radio_btn5.pack(padx=1, side='left')
#---
IPIP_22_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_22_frame.pack(fill = 'x')
IPIP_22_val = tk.StringVar()
IPIP_22_val.set('3')
IPIP_22_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_22_a.set("22.Am not interested in other people's problems.")    # 設定 a 的內容
IPIP_22_label = tk.Label(IPIP_22_frame, textvariable=IPIP_22_a, font=('Consolas',10))  # 放入標籤
IPIP_22_label.pack(side='left')
IPIP_22_label2 = tk.Label(IPIP_22_frame, textvariable=IPIP_22_val, font=('Consolas',10))  # 放入標籤
IPIP_22_label2.pack(side='left')
IPIP_22_radio_btn1 = tk.Radiobutton(IPIP_22_frame, text='1', variable=IPIP_22_val, value='1')    # 放入第一個單選按鈕
IPIP_22_radio_btn1.pack(padx=1, side='left')
IPIP_22_radio_btn2 = tk.Radiobutton(IPIP_22_frame, text='2', variable=IPIP_22_val, value='2')    # 放入第一個單選按鈕
IPIP_22_radio_btn2.pack(padx=1, side='left')
IPIP_22_radio_btn3 = tk.Radiobutton(IPIP_22_frame, text='3', variable=IPIP_22_val, value='3')    # 放入第一個單選按鈕
IPIP_22_radio_btn3.pack(padx=1, side='left')
IPIP_22_radio_btn4 = tk.Radiobutton(IPIP_22_frame, text='4', variable=IPIP_22_val, value='4')    # 放入第一個單選按鈕
IPIP_22_radio_btn4.pack(padx=1, side='left')
IPIP_22_radio_btn5 = tk.Radiobutton(IPIP_22_frame, text='5', variable=IPIP_22_val, value='5')    # 放入第一個單選按鈕
IPIP_22_radio_btn5.pack(padx=1, side='left')
#---
IPIP_23_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_23_frame.pack(fill = 'x')
IPIP_23_val = tk.StringVar()
IPIP_23_val.set('3')
IPIP_23_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_23_a.set("23.Get chores done right away.")    # 設定 a 的內容
IPIP_23_label = tk.Label(IPIP_23_frame, textvariable=IPIP_23_a, font=('Consolas',10))  # 放入標籤
IPIP_23_label.pack(side='left')
IPIP_23_label2 = tk.Label(IPIP_23_frame, textvariable=IPIP_23_val, font=('Consolas',10))  # 放入標籤
IPIP_23_label2.pack(side='left')
IPIP_23_radio_btn1 = tk.Radiobutton(IPIP_23_frame, text='1', variable=IPIP_23_val, value='1')    # 放入第一個單選按鈕
IPIP_23_radio_btn1.pack(padx=1, side='left')
IPIP_23_radio_btn2 = tk.Radiobutton(IPIP_23_frame, text='2', variable=IPIP_23_val, value='2')    # 放入第一個單選按鈕
IPIP_23_radio_btn2.pack(padx=1, side='left')
IPIP_23_radio_btn3 = tk.Radiobutton(IPIP_23_frame, text='3', variable=IPIP_23_val, value='3')    # 放入第一個單選按鈕
IPIP_23_radio_btn3.pack(padx=1, side='left')
IPIP_23_radio_btn4 = tk.Radiobutton(IPIP_23_frame, text='4', variable=IPIP_23_val, value='4')    # 放入第一個單選按鈕
IPIP_23_radio_btn4.pack(padx=1, side='left')
IPIP_23_radio_btn5 = tk.Radiobutton(IPIP_23_frame, text='5', variable=IPIP_23_val, value='5')    # 放入第一個單選按鈕
IPIP_23_radio_btn5.pack(padx=1, side='left')
#---
IPIP_24_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_24_frame.pack(fill = 'x')
IPIP_24_val = tk.StringVar()
IPIP_24_val.set('3')
IPIP_24_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_24_a.set("24.Am easily disturbed.")    # 設定 a 的內容
IPIP_24_label = tk.Label(IPIP_24_frame, textvariable=IPIP_24_a, font=('Consolas',10))  # 放入標籤
IPIP_24_label.pack(side='left')
IPIP_24_label2 = tk.Label(IPIP_24_frame, textvariable=IPIP_24_val, font=('Consolas',10))  # 放入標籤
IPIP_24_label2.pack(side='left')
IPIP_24_radio_btn1 = tk.Radiobutton(IPIP_24_frame, text='1', variable=IPIP_24_val, value='1')    # 放入第一個單選按鈕
IPIP_24_radio_btn1.pack(padx=1, side='left')
IPIP_24_radio_btn2 = tk.Radiobutton(IPIP_24_frame, text='2', variable=IPIP_24_val, value='2')    # 放入第一個單選按鈕
IPIP_24_radio_btn2.pack(padx=1, side='left')
IPIP_24_radio_btn3 = tk.Radiobutton(IPIP_24_frame, text='3', variable=IPIP_24_val, value='3')    # 放入第一個單選按鈕
IPIP_24_radio_btn3.pack(padx=1, side='left')
IPIP_24_radio_btn4 = tk.Radiobutton(IPIP_24_frame, text='4', variable=IPIP_24_val, value='4')    # 放入第一個單選按鈕
IPIP_24_radio_btn4.pack(padx=1, side='left')
IPIP_24_radio_btn5 = tk.Radiobutton(IPIP_24_frame, text='5', variable=IPIP_24_val, value='5')    # 放入第一個單選按鈕
IPIP_24_radio_btn5.pack(padx=1, side='left')
#---
IPIP_25_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_25_frame.pack(fill = 'x')
IPIP_25_val = tk.StringVar()
IPIP_25_val.set('3')
IPIP_25_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_25_a.set("25.Have excellent ideas.")    # 設定 a 的內容
IPIP_25_label = tk.Label(IPIP_25_frame, textvariable=IPIP_25_a, font=('Consolas',10))  # 放入標籤
IPIP_25_label.pack(side='left')
IPIP_25_label2 = tk.Label(IPIP_25_frame, textvariable=IPIP_25_val, font=('Consolas',10))  # 放入標籤
IPIP_25_label2.pack(side='left')
IPIP_25_radio_btn1 = tk.Radiobutton(IPIP_25_frame, text='1', variable=IPIP_25_val, value='1')    # 放入第一個單選按鈕
IPIP_25_radio_btn1.pack(padx=1, side='left')
IPIP_25_radio_btn2 = tk.Radiobutton(IPIP_25_frame, text='2', variable=IPIP_25_val, value='2')    # 放入第一個單選按鈕
IPIP_25_radio_btn2.pack(padx=1, side='left')
IPIP_25_radio_btn3 = tk.Radiobutton(IPIP_25_frame, text='3', variable=IPIP_25_val, value='3')    # 放入第一個單選按鈕
IPIP_25_radio_btn3.pack(padx=1, side='left')
IPIP_25_radio_btn4 = tk.Radiobutton(IPIP_25_frame, text='4', variable=IPIP_25_val, value='4')    # 放入第一個單選按鈕
IPIP_25_radio_btn4.pack(padx=1, side='left')
IPIP_25_radio_btn5 = tk.Radiobutton(IPIP_25_frame, text='5', variable=IPIP_25_val, value='5')    # 放入第一個單選按鈕
IPIP_25_radio_btn5.pack(padx=1, side='left')
#---
IPIP_26_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_26_frame.pack(fill = 'x')
IPIP_26_val = tk.StringVar()
IPIP_26_val.set('3')
IPIP_26_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_26_a.set("26.Have little to say.")    # 設定 a 的內容
IPIP_26_label = tk.Label(IPIP_26_frame, textvariable=IPIP_26_a, font=('Consolas',10))  # 放入標籤
IPIP_26_label.pack(side='left')
IPIP_26_label2 = tk.Label(IPIP_26_frame, textvariable=IPIP_26_val, font=('Consolas',10))  # 放入標籤
IPIP_26_label2.pack(side='left')
IPIP_26_radio_btn1 = tk.Radiobutton(IPIP_26_frame, text='1', variable=IPIP_26_val, value='1')    # 放入第一個單選按鈕
IPIP_26_radio_btn1.pack(padx=1, side='left')
IPIP_26_radio_btn2 = tk.Radiobutton(IPIP_26_frame, text='2', variable=IPIP_26_val, value='2')    # 放入第一個單選按鈕
IPIP_26_radio_btn2.pack(padx=1, side='left')
IPIP_26_radio_btn3 = tk.Radiobutton(IPIP_26_frame, text='3', variable=IPIP_26_val, value='3')    # 放入第一個單選按鈕
IPIP_26_radio_btn3.pack(padx=1, side='left')
IPIP_26_radio_btn4 = tk.Radiobutton(IPIP_26_frame, text='4', variable=IPIP_26_val, value='4')    # 放入第一個單選按鈕
IPIP_26_radio_btn4.pack(padx=1, side='left')
IPIP_26_radio_btn5 = tk.Radiobutton(IPIP_26_frame, text='5', variable=IPIP_26_val, value='5')    # 放入第一個單選按鈕
IPIP_26_radio_btn5.pack(padx=1, side='left')
#---
IPIP_27_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_27_frame.pack(fill = 'x')
IPIP_27_val = tk.StringVar()
IPIP_27_val.set('3')
IPIP_27_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_27_a.set("27.Have a soft heart.")    # 設定 a 的內容
IPIP_27_label = tk.Label(IPIP_27_frame, textvariable=IPIP_27_a, font=('Consolas',10))  # 放入標籤
IPIP_27_label.pack(side='left')
IPIP_27_label2 = tk.Label(IPIP_27_frame, textvariable=IPIP_27_val, font=('Consolas',10))  # 放入標籤
IPIP_27_label2.pack(side='left')
IPIP_27_radio_btn1 = tk.Radiobutton(IPIP_27_frame, text='1', variable=IPIP_27_val, value='1')    # 放入第一個單選按鈕
IPIP_27_radio_btn1.pack(padx=1, side='left')
IPIP_27_radio_btn2 = tk.Radiobutton(IPIP_27_frame, text='2', variable=IPIP_27_val, value='2')    # 放入第一個單選按鈕
IPIP_27_radio_btn2.pack(padx=1, side='left')
IPIP_27_radio_btn3 = tk.Radiobutton(IPIP_27_frame, text='3', variable=IPIP_27_val, value='3')    # 放入第一個單選按鈕
IPIP_27_radio_btn3.pack(padx=1, side='left')
IPIP_27_radio_btn4 = tk.Radiobutton(IPIP_27_frame, text='4', variable=IPIP_27_val, value='4')    # 放入第一個單選按鈕
IPIP_27_radio_btn4.pack(padx=1, side='left')
IPIP_27_radio_btn5 = tk.Radiobutton(IPIP_27_frame, text='5', variable=IPIP_27_val, value='5')    # 放入第一個單選按鈕
IPIP_27_radio_btn5.pack(padx=1, side='left')
#---
IPIP_28_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_28_frame.pack(fill = 'x')
IPIP_28_val = tk.StringVar()
IPIP_28_val.set('3')
IPIP_28_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_28_a.set("28.Often forget to put things back in their proper place.")    # 設定 a 的內容
IPIP_28_label = tk.Label(IPIP_28_frame, textvariable=IPIP_28_a, font=('Consolas',10))  # 放入標籤
IPIP_28_label.pack(side='left')
IPIP_28_label2 = tk.Label(IPIP_28_frame, textvariable=IPIP_28_val, font=('Consolas',10))  # 放入標籤
IPIP_28_label2.pack(side='left')
IPIP_28_radio_btn1 = tk.Radiobutton(IPIP_28_frame, text='1', variable=IPIP_28_val, value='1')    # 放入第一個單選按鈕
IPIP_28_radio_btn1.pack(padx=1, side='left')
IPIP_28_radio_btn2 = tk.Radiobutton(IPIP_28_frame, text='2', variable=IPIP_28_val, value='2')    # 放入第一個單選按鈕
IPIP_28_radio_btn2.pack(padx=1, side='left')
IPIP_28_radio_btn3 = tk.Radiobutton(IPIP_28_frame, text='3', variable=IPIP_28_val, value='3')    # 放入第一個單選按鈕
IPIP_28_radio_btn3.pack(padx=1, side='left')
IPIP_28_radio_btn4 = tk.Radiobutton(IPIP_28_frame, text='4', variable=IPIP_28_val, value='4')    # 放入第一個單選按鈕
IPIP_28_radio_btn4.pack(padx=1, side='left')
IPIP_28_radio_btn5 = tk.Radiobutton(IPIP_28_frame, text='5', variable=IPIP_28_val, value='5')    # 放入第一個單選按鈕
IPIP_28_radio_btn5.pack(padx=1, side='left')
#---
IPIP_29_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_29_frame.pack(fill = 'x')
IPIP_29_val = tk.StringVar()
IPIP_29_val.set('3')
IPIP_29_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_29_a.set("29.Get upset easily.")    # 設定 a 的內容
IPIP_29_label = tk.Label(IPIP_29_frame, textvariable=IPIP_29_a, font=('Consolas',10))  # 放入標籤
IPIP_29_label.pack(side='left')
IPIP_29_label2 = tk.Label(IPIP_29_frame, textvariable=IPIP_29_val, font=('Consolas',10))  # 放入標籤
IPIP_29_label2.pack(side='left')
IPIP_29_radio_btn1 = tk.Radiobutton(IPIP_29_frame, text='1', variable=IPIP_29_val, value='1')    # 放入第一個單選按鈕
IPIP_29_radio_btn1.pack(padx=1, side='left')
IPIP_29_radio_btn2 = tk.Radiobutton(IPIP_29_frame, text='2', variable=IPIP_29_val, value='2')    # 放入第一個單選按鈕
IPIP_29_radio_btn2.pack(padx=1, side='left')
IPIP_29_radio_btn3 = tk.Radiobutton(IPIP_29_frame, text='3', variable=IPIP_29_val, value='3')    # 放入第一個單選按鈕
IPIP_29_radio_btn3.pack(padx=1, side='left')
IPIP_29_radio_btn4 = tk.Radiobutton(IPIP_29_frame, text='4', variable=IPIP_29_val, value='4')    # 放入第一個單選按鈕
IPIP_29_radio_btn4.pack(padx=1, side='left')
IPIP_29_radio_btn5 = tk.Radiobutton(IPIP_29_frame, text='5', variable=IPIP_29_val, value='5')    # 放入第一個單選按鈕
IPIP_29_radio_btn5.pack(padx=1, side='left')
#---
IPIP_30_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_30_frame.pack(fill = 'x')
IPIP_30_val = tk.StringVar()
IPIP_30_val.set('3')
IPIP_30_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_30_a.set("30.Do not have a good imagination.")    # 設定 a 的內容
IPIP_30_label = tk.Label(IPIP_30_frame, textvariable=IPIP_30_a, font=('Consolas',10))  # 放入標籤
IPIP_30_label.pack(side='left')
IPIP_30_label2 = tk.Label(IPIP_30_frame, textvariable=IPIP_30_val, font=('Consolas',10))  # 放入標籤
IPIP_30_label2.pack(side='left')
IPIP_30_radio_btn1 = tk.Radiobutton(IPIP_30_frame, text='1', variable=IPIP_30_val, value='1')    # 放入第一個單選按鈕
IPIP_30_radio_btn1.pack(padx=1, side='left')
IPIP_30_radio_btn2 = tk.Radiobutton(IPIP_30_frame, text='2', variable=IPIP_30_val, value='2')    # 放入第一個單選按鈕
IPIP_30_radio_btn2.pack(padx=1, side='left')
IPIP_30_radio_btn3 = tk.Radiobutton(IPIP_30_frame, text='3', variable=IPIP_30_val, value='3')    # 放入第一個單選按鈕
IPIP_30_radio_btn3.pack(padx=1, side='left')
IPIP_30_radio_btn4 = tk.Radiobutton(IPIP_30_frame, text='4', variable=IPIP_30_val, value='4')    # 放入第一個單選按鈕
IPIP_30_radio_btn4.pack(padx=1, side='left')
IPIP_30_radio_btn5 = tk.Radiobutton(IPIP_30_frame, text='5', variable=IPIP_30_val, value='5')    # 放入第一個單選按鈕
IPIP_30_radio_btn5.pack(padx=1, side='left')
#---
IPIP_31_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_31_frame.pack(fill = 'x')
IPIP_31_val = tk.StringVar()
IPIP_31_val.set('3')
IPIP_31_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_31_a.set("31.Talk to a lot of different people at parties.")    # 設定 a 的內容
IPIP_31_label = tk.Label(IPIP_31_frame, textvariable=IPIP_31_a, font=('Consolas',10))  # 放入標籤
IPIP_31_label.pack(side='left')
IPIP_31_label2 = tk.Label(IPIP_31_frame, textvariable=IPIP_31_val, font=('Consolas',10))  # 放入標籤
IPIP_31_label2.pack(side='left')
IPIP_31_radio_btn1 = tk.Radiobutton(IPIP_31_frame, text='1', variable=IPIP_31_val, value='1')    # 放入第一個單選按鈕
IPIP_31_radio_btn1.pack(padx=1, side='left')
IPIP_31_radio_btn2 = tk.Radiobutton(IPIP_31_frame, text='2', variable=IPIP_31_val, value='2')    # 放入第一個單選按鈕
IPIP_31_radio_btn2.pack(padx=1, side='left')
IPIP_31_radio_btn3 = tk.Radiobutton(IPIP_31_frame, text='3', variable=IPIP_31_val, value='3')    # 放入第一個單選按鈕
IPIP_31_radio_btn3.pack(padx=1, side='left')
IPIP_31_radio_btn4 = tk.Radiobutton(IPIP_31_frame, text='4', variable=IPIP_31_val, value='4')    # 放入第一個單選按鈕
IPIP_31_radio_btn4.pack(padx=1, side='left')
IPIP_31_radio_btn5 = tk.Radiobutton(IPIP_31_frame, text='5', variable=IPIP_31_val, value='5')    # 放入第一個單選按鈕
IPIP_31_radio_btn5.pack(padx=1, side='left')
#---
IPIP_32_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_32_frame.pack(fill = 'x')
IPIP_32_val = tk.StringVar()
IPIP_32_val.set('3')
IPIP_32_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_32_a.set("32.Am not really interested in others.")    # 設定 a 的內容
IPIP_32_label = tk.Label(IPIP_32_frame, textvariable=IPIP_32_a, font=('Consolas',10))  # 放入標籤
IPIP_32_label.pack(side='left')
IPIP_32_label2 = tk.Label(IPIP_32_frame, textvariable=IPIP_32_val, font=('Consolas',10))  # 放入標籤
IPIP_32_label2.pack(side='left')
IPIP_32_radio_btn1 = tk.Radiobutton(IPIP_32_frame, text='1', variable=IPIP_32_val, value='1')    # 放入第一個單選按鈕
IPIP_32_radio_btn1.pack(padx=1, side='left')
IPIP_32_radio_btn2 = tk.Radiobutton(IPIP_32_frame, text='2', variable=IPIP_32_val, value='2')    # 放入第一個單選按鈕
IPIP_32_radio_btn2.pack(padx=1, side='left')
IPIP_32_radio_btn3 = tk.Radiobutton(IPIP_32_frame, text='3', variable=IPIP_32_val, value='3')    # 放入第一個單選按鈕
IPIP_32_radio_btn3.pack(padx=1, side='left')
IPIP_32_radio_btn4 = tk.Radiobutton(IPIP_32_frame, text='4', variable=IPIP_32_val, value='4')    # 放入第一個單選按鈕
IPIP_32_radio_btn4.pack(padx=1, side='left')
IPIP_32_radio_btn5 = tk.Radiobutton(IPIP_32_frame, text='5', variable=IPIP_32_val, value='5')    # 放入第一個單選按鈕
IPIP_32_radio_btn5.pack(padx=1, side='left')
#---
IPIP_33_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_33_frame.pack(fill = 'x')
IPIP_33_val = tk.StringVar()
IPIP_33_val.set('3')
IPIP_33_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_33_a.set("33.Like order.")    # 設定 a 的內容
IPIP_33_label = tk.Label(IPIP_33_frame, textvariable=IPIP_33_a, font=('Consolas',10))  # 放入標籤
IPIP_33_label.pack(side='left')
IPIP_33_label2 = tk.Label(IPIP_33_frame, textvariable=IPIP_33_val, font=('Consolas',10))  # 放入標籤
IPIP_33_label2.pack(side='left')
IPIP_33_radio_btn1 = tk.Radiobutton(IPIP_33_frame, text='1', variable=IPIP_33_val, value='1')    # 放入第一個單選按鈕
IPIP_33_radio_btn1.pack(padx=1, side='left')
IPIP_33_radio_btn2 = tk.Radiobutton(IPIP_33_frame, text='2', variable=IPIP_33_val, value='2')    # 放入第一個單選按鈕
IPIP_33_radio_btn2.pack(padx=1, side='left')
IPIP_33_radio_btn3 = tk.Radiobutton(IPIP_33_frame, text='3', variable=IPIP_33_val, value='3')    # 放入第一個單選按鈕
IPIP_33_radio_btn3.pack(padx=1, side='left')
IPIP_33_radio_btn4 = tk.Radiobutton(IPIP_33_frame, text='4', variable=IPIP_33_val, value='4')    # 放入第一個單選按鈕
IPIP_33_radio_btn4.pack(padx=1, side='left')
IPIP_33_radio_btn5 = tk.Radiobutton(IPIP_33_frame, text='5', variable=IPIP_33_val, value='5')    # 放入第一個單選按鈕
IPIP_33_radio_btn5.pack(padx=1, side='left')
#---
IPIP_34_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_34_frame.pack(fill = 'x')
IPIP_34_val = tk.StringVar()
IPIP_34_val.set('3')
IPIP_34_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_34_a.set("34.Change my mood a lot.")    # 設定 a 的內容
IPIP_34_label = tk.Label(IPIP_34_frame, textvariable=IPIP_34_a, font=('Consolas',10))  # 放入標籤
IPIP_34_label.pack(side='left')
IPIP_34_label2 = tk.Label(IPIP_34_frame, textvariable=IPIP_34_val, font=('Consolas',10))  # 放入標籤
IPIP_34_label2.pack(side='left')
IPIP_34_radio_btn1 = tk.Radiobutton(IPIP_34_frame, text='1', variable=IPIP_34_val, value='1')    # 放入第一個單選按鈕
IPIP_34_radio_btn1.pack(padx=1, side='left')
IPIP_34_radio_btn2 = tk.Radiobutton(IPIP_34_frame, text='2', variable=IPIP_34_val, value='2')    # 放入第一個單選按鈕
IPIP_34_radio_btn2.pack(padx=1, side='left')
IPIP_34_radio_btn3 = tk.Radiobutton(IPIP_34_frame, text='3', variable=IPIP_34_val, value='3')    # 放入第一個單選按鈕
IPIP_34_radio_btn3.pack(padx=1, side='left')
IPIP_34_radio_btn4 = tk.Radiobutton(IPIP_34_frame, text='4', variable=IPIP_34_val, value='4')    # 放入第一個單選按鈕
IPIP_34_radio_btn4.pack(padx=1, side='left')
IPIP_34_radio_btn5 = tk.Radiobutton(IPIP_34_frame, text='5', variable=IPIP_34_val, value='5')    # 放入第一個單選按鈕
IPIP_34_radio_btn5.pack(padx=1, side='left')
#---
IPIP_35_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_35_frame.pack(fill = 'x')
IPIP_35_val = tk.StringVar()
IPIP_35_val.set('3')
IPIP_35_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_35_a.set("35.Am quick to understand things.")    # 設定 a 的內容
IPIP_35_label = tk.Label(IPIP_35_frame, textvariable=IPIP_35_a, font=('Consolas',10))  # 放入標籤
IPIP_35_label.pack(side='left')
IPIP_35_label2 = tk.Label(IPIP_35_frame, textvariable=IPIP_35_val, font=('Consolas',10))  # 放入標籤
IPIP_35_label2.pack(side='left')
IPIP_35_radio_btn1 = tk.Radiobutton(IPIP_35_frame, text='1', variable=IPIP_35_val, value='1')    # 放入第一個單選按鈕
IPIP_35_radio_btn1.pack(padx=1, side='left')
IPIP_35_radio_btn2 = tk.Radiobutton(IPIP_35_frame, text='2', variable=IPIP_35_val, value='2')    # 放入第一個單選按鈕
IPIP_35_radio_btn2.pack(padx=1, side='left')
IPIP_35_radio_btn3 = tk.Radiobutton(IPIP_35_frame, text='3', variable=IPIP_35_val, value='3')    # 放入第一個單選按鈕
IPIP_35_radio_btn3.pack(padx=1, side='left')
IPIP_35_radio_btn4 = tk.Radiobutton(IPIP_35_frame, text='4', variable=IPIP_35_val, value='4')    # 放入第一個單選按鈕
IPIP_35_radio_btn4.pack(padx=1, side='left')
IPIP_35_radio_btn5 = tk.Radiobutton(IPIP_35_frame, text='5', variable=IPIP_35_val, value='5')    # 放入第一個單選按鈕
IPIP_35_radio_btn5.pack(padx=1, side='left')
#---
IPIP_36_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_36_frame.pack(fill = 'x')
IPIP_36_val = tk.StringVar()
IPIP_36_val.set('3')
IPIP_36_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_36_a.set("36.Don't like to draw attention to myself.")    # 設定 a 的內容
IPIP_36_label = tk.Label(IPIP_36_frame, textvariable=IPIP_36_a, font=('Consolas',10))  # 放入標籤
IPIP_36_label.pack(side='left')
IPIP_36_label2 = tk.Label(IPIP_36_frame, textvariable=IPIP_36_val, font=('Consolas',10))  # 放入標籤
IPIP_36_label2.pack(side='left')
IPIP_36_radio_btn1 = tk.Radiobutton(IPIP_36_frame, text='1', variable=IPIP_36_val, value='1')    # 放入第一個單選按鈕
IPIP_36_radio_btn1.pack(padx=1, side='left')
IPIP_36_radio_btn2 = tk.Radiobutton(IPIP_36_frame, text='2', variable=IPIP_36_val, value='2')    # 放入第一個單選按鈕
IPIP_36_radio_btn2.pack(padx=1, side='left')
IPIP_36_radio_btn3 = tk.Radiobutton(IPIP_36_frame, text='3', variable=IPIP_36_val, value='3')    # 放入第一個單選按鈕
IPIP_36_radio_btn3.pack(padx=1, side='left')
IPIP_36_radio_btn4 = tk.Radiobutton(IPIP_36_frame, text='4', variable=IPIP_36_val, value='4')    # 放入第一個單選按鈕
IPIP_36_radio_btn4.pack(padx=1, side='left')
IPIP_36_radio_btn5 = tk.Radiobutton(IPIP_36_frame, text='5', variable=IPIP_36_val, value='5')    # 放入第一個單選按鈕
IPIP_36_radio_btn5.pack(padx=1, side='left')
#---
IPIP_37_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_37_frame.pack(fill = 'x')
IPIP_37_val = tk.StringVar()
IPIP_37_val.set('3')
IPIP_37_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_37_a.set("37.Take time out for others.")    # 設定 a 的內容
IPIP_37_label = tk.Label(IPIP_37_frame, textvariable=IPIP_37_a, font=('Consolas',10))  # 放入標籤
IPIP_37_label.pack(side='left')
IPIP_37_label2 = tk.Label(IPIP_37_frame, textvariable=IPIP_37_val, font=('Consolas',10))  # 放入標籤
IPIP_37_label2.pack(side='left')
IPIP_37_radio_btn1 = tk.Radiobutton(IPIP_37_frame, text='1', variable=IPIP_37_val, value='1')    # 放入第一個單選按鈕
IPIP_37_radio_btn1.pack(padx=1, side='left')
IPIP_37_radio_btn2 = tk.Radiobutton(IPIP_37_frame, text='2', variable=IPIP_37_val, value='2')    # 放入第一個單選按鈕
IPIP_37_radio_btn2.pack(padx=1, side='left')
IPIP_37_radio_btn3 = tk.Radiobutton(IPIP_37_frame, text='3', variable=IPIP_37_val, value='3')    # 放入第一個單選按鈕
IPIP_37_radio_btn3.pack(padx=1, side='left')
IPIP_37_radio_btn4 = tk.Radiobutton(IPIP_37_frame, text='4', variable=IPIP_37_val, value='4')    # 放入第一個單選按鈕
IPIP_37_radio_btn4.pack(padx=1, side='left')
IPIP_37_radio_btn5 = tk.Radiobutton(IPIP_37_frame, text='5', variable=IPIP_37_val, value='5')    # 放入第一個單選按鈕
IPIP_37_radio_btn5.pack(padx=1, side='left')
#---
IPIP_38_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_38_frame.pack(fill = 'x')
IPIP_38_val = tk.StringVar()
IPIP_38_val.set('3')
IPIP_38_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_38_a.set("38.Shirk my duties.")    # 設定 a 的內容
IPIP_38_label = tk.Label(IPIP_38_frame, textvariable=IPIP_38_a, font=('Consolas',10))  # 放入標籤
IPIP_38_label.pack(side='left')
IPIP_38_label2 = tk.Label(IPIP_38_frame, textvariable=IPIP_38_val, font=('Consolas',10))  # 放入標籤
IPIP_38_label2.pack(side='left')
IPIP_38_radio_btn1 = tk.Radiobutton(IPIP_38_frame, text='1', variable=IPIP_38_val, value='1')    # 放入第一個單選按鈕
IPIP_38_radio_btn1.pack(padx=1, side='left')
IPIP_38_radio_btn2 = tk.Radiobutton(IPIP_38_frame, text='2', variable=IPIP_38_val, value='2')    # 放入第一個單選按鈕
IPIP_38_radio_btn2.pack(padx=1, side='left')
IPIP_38_radio_btn3 = tk.Radiobutton(IPIP_38_frame, text='3', variable=IPIP_38_val, value='3')    # 放入第一個單選按鈕
IPIP_38_radio_btn3.pack(padx=1, side='left')
IPIP_38_radio_btn4 = tk.Radiobutton(IPIP_38_frame, text='4', variable=IPIP_38_val, value='4')    # 放入第一個單選按鈕
IPIP_38_radio_btn4.pack(padx=1, side='left')
IPIP_38_radio_btn5 = tk.Radiobutton(IPIP_38_frame, text='5', variable=IPIP_38_val, value='5')    # 放入第一個單選按鈕
IPIP_38_radio_btn5.pack(padx=1, side='left')
#---
IPIP_39_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_39_frame.pack(fill = 'x')
IPIP_39_val = tk.StringVar()
IPIP_39_val.set('3')
IPIP_39_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_39_a.set("39.Have frequent mood swings.")    # 設定 a 的內容
IPIP_39_label = tk.Label(IPIP_39_frame, textvariable=IPIP_39_a, font=('Consolas',10))  # 放入標籤
IPIP_39_label.pack(side='left')
IPIP_39_label2 = tk.Label(IPIP_39_frame, textvariable=IPIP_39_val, font=('Consolas',10))  # 放入標籤
IPIP_39_label2.pack(side='left')
IPIP_39_radio_btn1 = tk.Radiobutton(IPIP_39_frame, text='1', variable=IPIP_39_val, value='1')    # 放入第一個單選按鈕
IPIP_39_radio_btn1.pack(padx=1, side='left')
IPIP_39_radio_btn2 = tk.Radiobutton(IPIP_39_frame, text='2', variable=IPIP_39_val, value='2')    # 放入第一個單選按鈕
IPIP_39_radio_btn2.pack(padx=1, side='left')
IPIP_39_radio_btn3 = tk.Radiobutton(IPIP_39_frame, text='3', variable=IPIP_39_val, value='3')    # 放入第一個單選按鈕
IPIP_39_radio_btn3.pack(padx=1, side='left')
IPIP_39_radio_btn4 = tk.Radiobutton(IPIP_39_frame, text='4', variable=IPIP_39_val, value='4')    # 放入第一個單選按鈕
IPIP_39_radio_btn4.pack(padx=1, side='left')
IPIP_39_radio_btn5 = tk.Radiobutton(IPIP_39_frame, text='5', variable=IPIP_39_val, value='5')    # 放入第一個單選按鈕
IPIP_39_radio_btn5.pack(padx=1, side='left')
#---
IPIP_40_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_40_frame.pack(fill = 'x')
IPIP_40_val = tk.StringVar()
IPIP_40_val.set('3')
IPIP_40_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_40_a.set("40.Use difficult words.")    # 設定 a 的內容
IPIP_40_label = tk.Label(IPIP_40_frame, textvariable=IPIP_40_a, font=('Consolas',10))  # 放入標籤
IPIP_40_label.pack(side='left')
IPIP_40_label2 = tk.Label(IPIP_40_frame, textvariable=IPIP_40_val, font=('Consolas',10))  # 放入標籤
IPIP_40_label2.pack(side='left')
IPIP_40_radio_btn1 = tk.Radiobutton(IPIP_40_frame, text='1', variable=IPIP_40_val, value='1')    # 放入第一個單選按鈕
IPIP_40_radio_btn1.pack(padx=1, side='left')
IPIP_40_radio_btn2 = tk.Radiobutton(IPIP_40_frame, text='2', variable=IPIP_40_val, value='2')    # 放入第一個單選按鈕
IPIP_40_radio_btn2.pack(padx=1, side='left')
IPIP_40_radio_btn3 = tk.Radiobutton(IPIP_40_frame, text='3', variable=IPIP_40_val, value='3')    # 放入第一個單選按鈕
IPIP_40_radio_btn3.pack(padx=1, side='left')
IPIP_40_radio_btn4 = tk.Radiobutton(IPIP_40_frame, text='4', variable=IPIP_40_val, value='4')    # 放入第一個單選按鈕
IPIP_40_radio_btn4.pack(padx=1, side='left')
IPIP_40_radio_btn5 = tk.Radiobutton(IPIP_40_frame, text='5', variable=IPIP_40_val, value='5')    # 放入第一個單選按鈕
IPIP_40_radio_btn5.pack(padx=1, side='left')
#---
IPIP_41_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_41_frame.pack(fill = 'x')
IPIP_41_val = tk.StringVar()
IPIP_41_val.set('3')
IPIP_41_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_41_a.set("41.Don't mind being the center of attention.")    # 設定 a 的內容
IPIP_41_label = tk.Label(IPIP_41_frame, textvariable=IPIP_41_a, font=('Consolas',10))  # 放入標籤
IPIP_41_label.pack(side='left')
IPIP_41_label2 = tk.Label(IPIP_41_frame, textvariable=IPIP_41_val, font=('Consolas',10))  # 放入標籤
IPIP_41_label2.pack(side='left')
IPIP_41_radio_btn1 = tk.Radiobutton(IPIP_41_frame, text='1', variable=IPIP_41_val, value='1')    # 放入第一個單選按鈕
IPIP_41_radio_btn1.pack(padx=1, side='left')
IPIP_41_radio_btn2 = tk.Radiobutton(IPIP_41_frame, text='2', variable=IPIP_41_val, value='2')    # 放入第一個單選按鈕
IPIP_41_radio_btn2.pack(padx=1, side='left')
IPIP_41_radio_btn3 = tk.Radiobutton(IPIP_41_frame, text='3', variable=IPIP_41_val, value='3')    # 放入第一個單選按鈕
IPIP_41_radio_btn3.pack(padx=1, side='left')
IPIP_41_radio_btn4 = tk.Radiobutton(IPIP_41_frame, text='4', variable=IPIP_41_val, value='4')    # 放入第一個單選按鈕
IPIP_41_radio_btn4.pack(padx=1, side='left')
IPIP_41_radio_btn5 = tk.Radiobutton(IPIP_41_frame, text='5', variable=IPIP_41_val, value='5')    # 放入第一個單選按鈕
IPIP_41_radio_btn5.pack(padx=1, side='left')
#---
IPIP_42_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_42_frame.pack(fill = 'x')
IPIP_42_val = tk.StringVar()
IPIP_42_val.set('3')
IPIP_42_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_42_a.set("42.Feel others' emotions.")    # 設定 a 的內容
IPIP_42_label = tk.Label(IPIP_42_frame, textvariable=IPIP_42_a, font=('Consolas',10))  # 放入標籤
IPIP_42_label.pack(side='left')
IPIP_42_label2 = tk.Label(IPIP_42_frame, textvariable=IPIP_42_val, font=('Consolas',10))  # 放入標籤
IPIP_42_label2.pack(side='left')
IPIP_42_radio_btn1 = tk.Radiobutton(IPIP_42_frame, text='1', variable=IPIP_42_val, value='1')    # 放入第一個單選按鈕
IPIP_42_radio_btn1.pack(padx=1, side='left')
IPIP_42_radio_btn2 = tk.Radiobutton(IPIP_42_frame, text='2', variable=IPIP_42_val, value='2')    # 放入第一個單選按鈕
IPIP_42_radio_btn2.pack(padx=1, side='left')
IPIP_42_radio_btn3 = tk.Radiobutton(IPIP_42_frame, text='3', variable=IPIP_42_val, value='3')    # 放入第一個單選按鈕
IPIP_42_radio_btn3.pack(padx=1, side='left')
IPIP_42_radio_btn4 = tk.Radiobutton(IPIP_42_frame, text='4', variable=IPIP_42_val, value='4')    # 放入第一個單選按鈕
IPIP_42_radio_btn4.pack(padx=1, side='left')
IPIP_42_radio_btn5 = tk.Radiobutton(IPIP_42_frame, text='5', variable=IPIP_42_val, value='5')    # 放入第一個單選按鈕
IPIP_42_radio_btn5.pack(padx=1, side='left')
#---
IPIP_43_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_43_frame.pack(fill = 'x')
IPIP_43_val = tk.StringVar()
IPIP_43_val.set('3')
IPIP_43_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_43_a.set("43.Follow a schedule.")    # 設定 a 的內容
IPIP_43_label = tk.Label(IPIP_43_frame, textvariable=IPIP_43_a, font=('Consolas',10))  # 放入標籤
IPIP_43_label.pack(side='left')
IPIP_43_label2 = tk.Label(IPIP_43_frame, textvariable=IPIP_43_val, font=('Consolas',10))  # 放入標籤
IPIP_43_label2.pack(side='left')
IPIP_43_radio_btn1 = tk.Radiobutton(IPIP_43_frame, text='1', variable=IPIP_43_val, value='1')    # 放入第一個單選按鈕
IPIP_43_radio_btn1.pack(padx=1, side='left')
IPIP_43_radio_btn2 = tk.Radiobutton(IPIP_43_frame, text='2', variable=IPIP_43_val, value='2')    # 放入第一個單選按鈕
IPIP_43_radio_btn2.pack(padx=1, side='left')
IPIP_43_radio_btn3 = tk.Radiobutton(IPIP_43_frame, text='3', variable=IPIP_43_val, value='3')    # 放入第一個單選按鈕
IPIP_43_radio_btn3.pack(padx=1, side='left')
IPIP_43_radio_btn4 = tk.Radiobutton(IPIP_43_frame, text='4', variable=IPIP_43_val, value='4')    # 放入第一個單選按鈕
IPIP_43_radio_btn4.pack(padx=1, side='left')
IPIP_43_radio_btn5 = tk.Radiobutton(IPIP_43_frame, text='5', variable=IPIP_43_val, value='5')    # 放入第一個單選按鈕
IPIP_43_radio_btn5.pack(padx=1, side='left')
#---
IPIP_44_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_44_frame.pack(fill = 'x')
IPIP_44_val = tk.StringVar()
IPIP_44_val.set('3')
IPIP_44_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_44_a.set("44.Get irritated easily.")    # 設定 a 的內容
IPIP_44_label = tk.Label(IPIP_44_frame, textvariable=IPIP_44_a, font=('Consolas',10))  # 放入標籤
IPIP_44_label.pack(side='left')
IPIP_44_label2 = tk.Label(IPIP_44_frame, textvariable=IPIP_44_val, font=('Consolas',10))  # 放入標籤
IPIP_44_label2.pack(side='left')
IPIP_44_radio_btn1 = tk.Radiobutton(IPIP_44_frame, text='1', variable=IPIP_44_val, value='1')    # 放入第一個單選按鈕
IPIP_44_radio_btn1.pack(padx=1, side='left')
IPIP_44_radio_btn2 = tk.Radiobutton(IPIP_44_frame, text='2', variable=IPIP_44_val, value='2')    # 放入第一個單選按鈕
IPIP_44_radio_btn2.pack(padx=1, side='left')
IPIP_44_radio_btn3 = tk.Radiobutton(IPIP_44_frame, text='3', variable=IPIP_44_val, value='3')    # 放入第一個單選按鈕
IPIP_44_radio_btn3.pack(padx=1, side='left')
IPIP_44_radio_btn4 = tk.Radiobutton(IPIP_44_frame, text='4', variable=IPIP_44_val, value='4')    # 放入第一個單選按鈕
IPIP_44_radio_btn4.pack(padx=1, side='left')
IPIP_44_radio_btn5 = tk.Radiobutton(IPIP_44_frame, text='5', variable=IPIP_44_val, value='5')    # 放入第一個單選按鈕
IPIP_44_radio_btn5.pack(padx=1, side='left')
#---
IPIP_45_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_45_frame.pack(fill = 'x')
IPIP_45_val = tk.StringVar()
IPIP_45_val.set('3')
IPIP_45_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_45_a.set("45.Spend time reflecting on things.")    # 設定 a 的內容
IPIP_45_label = tk.Label(IPIP_45_frame, textvariable=IPIP_45_a, font=('Consolas',10))  # 放入標籤
IPIP_45_label.pack(side='left')
IPIP_45_label2 = tk.Label(IPIP_45_frame, textvariable=IPIP_45_val, font=('Consolas',10))  # 放入標籤
IPIP_45_label2.pack(side='left')
IPIP_45_radio_btn1 = tk.Radiobutton(IPIP_45_frame, text='1', variable=IPIP_45_val, value='1')    # 放入第一個單選按鈕
IPIP_45_radio_btn1.pack(padx=1, side='left')
IPIP_45_radio_btn2 = tk.Radiobutton(IPIP_45_frame, text='2', variable=IPIP_45_val, value='2')    # 放入第一個單選按鈕
IPIP_45_radio_btn2.pack(padx=1, side='left')
IPIP_45_radio_btn3 = tk.Radiobutton(IPIP_45_frame, text='3', variable=IPIP_45_val, value='3')    # 放入第一個單選按鈕
IPIP_45_radio_btn3.pack(padx=1, side='left')
IPIP_45_radio_btn4 = tk.Radiobutton(IPIP_45_frame, text='4', variable=IPIP_45_val, value='4')    # 放入第一個單選按鈕
IPIP_45_radio_btn4.pack(padx=1, side='left')
IPIP_45_radio_btn5 = tk.Radiobutton(IPIP_45_frame, text='5', variable=IPIP_45_val, value='5')    # 放入第一個單選按鈕
IPIP_45_radio_btn5.pack(padx=1, side='left')
#---
IPIP_46_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_46_frame.pack(fill = 'x')
IPIP_46_val = tk.StringVar()
IPIP_46_val.set('3')
IPIP_46_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_46_a.set("46.Am quiet around strangers.")    # 設定 a 的內容
IPIP_46_label = tk.Label(IPIP_46_frame, textvariable=IPIP_46_a, font=('Consolas',10))  # 放入標籤
IPIP_46_label.pack(side='left')
IPIP_46_label2 = tk.Label(IPIP_46_frame, textvariable=IPIP_46_val, font=('Consolas',10))  # 放入標籤
IPIP_46_label2.pack(side='left')
IPIP_46_radio_btn1 = tk.Radiobutton(IPIP_46_frame, text='1', variable=IPIP_46_val, value='1')    # 放入第一個單選按鈕
IPIP_46_radio_btn1.pack(padx=1, side='left')
IPIP_46_radio_btn2 = tk.Radiobutton(IPIP_46_frame, text='2', variable=IPIP_46_val, value='2')    # 放入第一個單選按鈕
IPIP_46_radio_btn2.pack(padx=1, side='left')
IPIP_46_radio_btn3 = tk.Radiobutton(IPIP_46_frame, text='3', variable=IPIP_46_val, value='3')    # 放入第一個單選按鈕
IPIP_46_radio_btn3.pack(padx=1, side='left')
IPIP_46_radio_btn4 = tk.Radiobutton(IPIP_46_frame, text='4', variable=IPIP_46_val, value='4')    # 放入第一個單選按鈕
IPIP_46_radio_btn4.pack(padx=1, side='left')
IPIP_46_radio_btn5 = tk.Radiobutton(IPIP_46_frame, text='5', variable=IPIP_46_val, value='5')    # 放入第一個單選按鈕
IPIP_46_radio_btn5.pack(padx=1, side='left')
#---
IPIP_47_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_47_frame.pack(fill = 'x')
IPIP_47_val = tk.StringVar()
IPIP_47_val.set('3')
IPIP_47_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_47_a.set("47.Make people feel at ease.")    # 設定 a 的內容
IPIP_47_label = tk.Label(IPIP_47_frame, textvariable=IPIP_47_a, font=('Consolas',10))  # 放入標籤
IPIP_47_label.pack(side='left')
IPIP_47_label2 = tk.Label(IPIP_47_frame, textvariable=IPIP_47_val, font=('Consolas',10))  # 放入標籤
IPIP_47_label2.pack(side='left')
IPIP_47_radio_btn1 = tk.Radiobutton(IPIP_47_frame, text='1', variable=IPIP_47_val, value='1')    # 放入第一個單選按鈕
IPIP_47_radio_btn1.pack(padx=1, side='left')
IPIP_47_radio_btn2 = tk.Radiobutton(IPIP_47_frame, text='2', variable=IPIP_47_val, value='2')    # 放入第一個單選按鈕
IPIP_47_radio_btn2.pack(padx=1, side='left')
IPIP_47_radio_btn3 = tk.Radiobutton(IPIP_47_frame, text='3', variable=IPIP_47_val, value='3')    # 放入第一個單選按鈕
IPIP_47_radio_btn3.pack(padx=1, side='left')
IPIP_47_radio_btn4 = tk.Radiobutton(IPIP_47_frame, text='4', variable=IPIP_47_val, value='4')    # 放入第一個單選按鈕
IPIP_47_radio_btn4.pack(padx=1, side='left')
IPIP_47_radio_btn5 = tk.Radiobutton(IPIP_47_frame, text='5', variable=IPIP_47_val, value='5')    # 放入第一個單選按鈕
IPIP_47_radio_btn5.pack(padx=1, side='left')
#---
IPIP_48_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_48_frame.pack(fill = 'x')
IPIP_48_val = tk.StringVar()
IPIP_48_val.set('3')
IPIP_48_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_48_a.set("48.Am exacting in my work.")    # 設定 a 的內容
IPIP_48_label = tk.Label(IPIP_48_frame, textvariable=IPIP_48_a, font=('Consolas',10))  # 放入標籤
IPIP_48_label.pack(side='left')
IPIP_48_label2 = tk.Label(IPIP_48_frame, textvariable=IPIP_48_val, font=('Consolas',10))  # 放入標籤
IPIP_48_label2.pack(side='left')
IPIP_48_radio_btn1 = tk.Radiobutton(IPIP_48_frame, text='1', variable=IPIP_48_val, value='1')    # 放入第一個單選按鈕
IPIP_48_radio_btn1.pack(padx=1, side='left')
IPIP_48_radio_btn2 = tk.Radiobutton(IPIP_48_frame, text='2', variable=IPIP_48_val, value='2')    # 放入第一個單選按鈕
IPIP_48_radio_btn2.pack(padx=1, side='left')
IPIP_48_radio_btn3 = tk.Radiobutton(IPIP_48_frame, text='3', variable=IPIP_48_val, value='3')    # 放入第一個單選按鈕
IPIP_48_radio_btn3.pack(padx=1, side='left')
IPIP_48_radio_btn4 = tk.Radiobutton(IPIP_48_frame, text='4', variable=IPIP_48_val, value='4')    # 放入第一個單選按鈕
IPIP_48_radio_btn4.pack(padx=1, side='left')
IPIP_48_radio_btn5 = tk.Radiobutton(IPIP_48_frame, text='5', variable=IPIP_48_val, value='5')    # 放入第一個單選按鈕
IPIP_48_radio_btn5.pack(padx=1, side='left')
#---
IPIP_49_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_49_frame.pack(fill = 'x')
IPIP_49_val = tk.StringVar()
IPIP_49_val.set('3')
IPIP_49_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_49_a.set("49.Often feel blue.")    # 設定 a 的內容
IPIP_49_label = tk.Label(IPIP_49_frame, textvariable=IPIP_49_a, font=('Consolas',10))  # 放入標籤
IPIP_49_label.pack(side='left')
IPIP_49_label2 = tk.Label(IPIP_49_frame, textvariable=IPIP_49_val, font=('Consolas',10))  # 放入標籤
IPIP_49_label2.pack(side='left')
IPIP_49_radio_btn1 = tk.Radiobutton(IPIP_49_frame, text='1', variable=IPIP_49_val, value='1')    # 放入第一個單選按鈕
IPIP_49_radio_btn1.pack(padx=1, side='left')
IPIP_49_radio_btn2 = tk.Radiobutton(IPIP_49_frame, text='2', variable=IPIP_49_val, value='2')    # 放入第一個單選按鈕
IPIP_49_radio_btn2.pack(padx=1, side='left')
IPIP_49_radio_btn3 = tk.Radiobutton(IPIP_49_frame, text='3', variable=IPIP_49_val, value='3')    # 放入第一個單選按鈕
IPIP_49_radio_btn3.pack(padx=1, side='left')
IPIP_49_radio_btn4 = tk.Radiobutton(IPIP_49_frame, text='4', variable=IPIP_49_val, value='4')    # 放入第一個單選按鈕
IPIP_49_radio_btn4.pack(padx=1, side='left')
IPIP_49_radio_btn5 = tk.Radiobutton(IPIP_49_frame, text='5', variable=IPIP_49_val, value='5')    # 放入第一個單選按鈕
IPIP_49_radio_btn5.pack(padx=1, side='left')
#---
IPIP_50_frame = tk.Frame(IPIP_frame, pady=5)                  # 加入 Frame 框架
IPIP_50_frame.pack(fill = 'x')
IPIP_50_val = tk.StringVar()
IPIP_50_val.set('3')
IPIP_50_a = tk.StringVar()  # 設定 a 為文字變數
IPIP_50_a.set("50.Am full of ideas.")    # 設定 a 的內容
IPIP_50_label = tk.Label(IPIP_50_frame, textvariable=IPIP_50_a, font=('Consolas',10))  # 放入標籤
IPIP_50_label.pack(side='left')
IPIP_50_label2 = tk.Label(IPIP_50_frame, textvariable=IPIP_50_val, font=('Consolas',10))  # 放入標籤
IPIP_50_label2.pack(side='left')
IPIP_50_radio_btn1 = tk.Radiobutton(IPIP_50_frame, text='1', variable=IPIP_50_val, value='1')    # 放入第一個單選按鈕
IPIP_50_radio_btn1.pack(padx=1, side='left')
IPIP_50_radio_btn2 = tk.Radiobutton(IPIP_50_frame, text='2', variable=IPIP_50_val, value='2')    # 放入第一個單選按鈕
IPIP_50_radio_btn2.pack(padx=1, side='left')
IPIP_50_radio_btn3 = tk.Radiobutton(IPIP_50_frame, text='3', variable=IPIP_50_val, value='3')    # 放入第一個單選按鈕
IPIP_50_radio_btn3.pack(padx=1, side='left')
IPIP_50_radio_btn4 = tk.Radiobutton(IPIP_50_frame, text='4', variable=IPIP_50_val, value='4')    # 放入第一個單選按鈕
IPIP_50_radio_btn4.pack(padx=1, side='left')
IPIP_50_radio_btn5 = tk.Radiobutton(IPIP_50_frame, text='5', variable=IPIP_50_val, value='5')    # 放入第一個單選按鈕
IPIP_50_radio_btn5.pack(padx=1, side='left')
#------------------------------------------------------------------------------
def pop_up():
    messagebox.showinfo("Wizarding World", house_str)

submitbutton = tk.Button(total_frame.interior, pady=5, bg='green', text='Finish', width='10', height='2', command= pop_up)
submitbutton.pack()
#--------------------------------------------------------------------------------
# print(SD3_M1_val.get())
SD3_Machiavellianism = int(SD3_M1_val.get()) + int(SD3_M2_val.get()) + int(SD3_M3_val.get()) + int(SD3_M4_val.get()) + int(SD3_M5_val.get()) + int(SD3_M6_val.get()) + int(SD3_M7_val.get()) + int(SD3_M8_val.get()) + int(SD3_M9_val.get())
SD3_Narcissism = int(SD3_N1_val.get()) + int(SD3_N2_val.get()) + int(SD3_N3_val.get()) + int(SD3_N4_val.get()) + int(SD3_N5_val.get()) + int(SD3_N6_val.get()) + int(SD3_N7_val.get()) + int(SD3_N8_val.get()) + int(SD3_N9_val.get())
SD3_Psychopathy = int(SD3_P1_val.get()) + int(SD3_P2_val.get()) + int(SD3_P3_val.get()) + int(SD3_P4_val.get()) + int(SD3_P5_val.get()) + int(SD3_P6_val.get()) + int(SD3_P7_val.get()) + int(SD3_P8_val.get()) + int(SD3_P9_val.get())
IPIP_Extraversion = int(IPIP_01_val.get()) - int(IPIP_06_val.get()) + int(IPIP_11_val.get()) - int(IPIP_16_val.get()) + int(IPIP_21_val.get()) - int(IPIP_26_val.get()) + int(IPIP_31_val.get()) - int(IPIP_36_val.get()) + int(IPIP_41_val.get()) - int(IPIP_46_val.get())
IPIP_Agreeableness = -int(IPIP_02_val.get()) + int(IPIP_07_val.get()) - int(IPIP_12_val.get()) + int(IPIP_17_val.get()) - int(IPIP_22_val.get()) + int(IPIP_27_val.get()) - int(IPIP_32_val.get()) + int(IPIP_37_val.get()) + int(IPIP_42_val.get()) + int(IPIP_47_val.get())
IPIP_Conscientiousness = int(IPIP_03_val.get()) - int(IPIP_08_val.get()) + int(IPIP_13_val.get()) - int(IPIP_18_val.get()) + int(IPIP_23_val.get()) - int(IPIP_28_val.get()) + int(IPIP_33_val.get()) - int(IPIP_38_val.get()) + int(IPIP_43_val.get()) + int(IPIP_48_val.get())
IPIP_EmStability = -int(IPIP_04_val.get()) + int(IPIP_09_val.get()) - int(IPIP_14_val.get()) + int(IPIP_19_val.get()) - int(IPIP_24_val.get()) - int(IPIP_29_val.get()) - int(IPIP_34_val.get()) - int(IPIP_39_val.get()) - int(IPIP_44_val.get()) - int(IPIP_49_val.get())
IPIP_Intellect = int(IPIP_05_val.get()) - int(IPIP_10_val.get()) + int(IPIP_15_val.get()) - int(IPIP_20_val.get()) + int(IPIP_25_val.get()) - int(IPIP_30_val.get()) + int(IPIP_35_val.get()) + int(IPIP_40_val.get()) + int(IPIP_45_val.get()) + int(IPIP_50_val.get())
if Sorting_house_wish_value.get() == 'Gryffindor':
  Sorting_house_wish = 1
elif Sorting_house_wish_value.get() == 'Ravenclaw':
  Sorting_house_wish = 2
elif Sorting_house_wish_value.get() == 'Hufflepuff':
  Sorting_house_wish = 3
else: #Sorting_house_wish == 'Slytherin'
  Sorting_house_wish = 4

if occupation_value.get() == 'Student':
  occupation = 1
elif occupation_value.get() == 'Employed':
  occupation = 2
elif occupation_value.get() == 'Unemployed':
  occupation = 3
else:
  occupation = 4

if gender_value.get() == 'Female':
  gender = 1
elif gender_value.get() == 'Male':
  gender = 2
X_test = np.array([Sorting_house_wish, age, occupation, gender,
       SD3_Machiavellianism, SD3_Narcissism, SD3_Psychopathy,
       IPIP_Extraversion, IPIP_Agreeableness, IPIP_Conscientiousness,
       IPIP_EmStability, IPIP_Intellect]).reshape(1, -1)
# model_entropy = DecisionTreeClassifier(criterion = 'entropy', random_state = 42, max_depth = 3, min_samples_leaf = 5)
# model_entropy = model_entropy.fit(X_train, Y_train)
Y_test_entropy = model_entropy.predict(X_test)
ans = int(Y_test_entropy)
if ans/3 >= 3.5:
    house_str='WELCOME TO SLYTHERIN!'
elif ans/3 >= 2.5:
    house_str='WELCOME TO HUFFLEPUFF!'
elif ans/3 >= 1.5:
    house_str='WELCOME TO RAVENCLAW!'
else: #ans/3 >= 0.5:
    house_str='WELCOME TO GRYFFINDOR!'
root.mainloop()  # 放在主迴圈中
```
